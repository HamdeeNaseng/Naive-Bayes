**‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á**: ‡∏ñ‡πâ‡∏≤ Œµ = 0.1
- 10% ‡∏Ç‡∏≠‡∏á‡πÄ‡∏ß‡∏•‡∏≤ ‚Üí ‡∏•‡∏≠‡∏á‡πÄ‡∏•‡∏∑‡∏≠‡∏Å‡πÅ‡∏ö‡∏ö‡∏™‡∏∏‡πà‡∏° (‡∏≠‡∏≤‡∏à‡πÑ‡∏î‡πâ arm ‡∏ó‡∏µ‡πà‡πÅ‡∏¢‡πà)
- 90% ‡∏Ç‡∏≠‡∏á‡πÄ‡∏ß‡∏•‡∏≤ ‚Üí ‡πÄ‡∏•‡∏∑‡∏≠‡∏Å arm ‡∏ó‡∏µ‡πà‡∏î‡∏µ‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î‡∏ó‡∏µ‡πà‡∏£‡∏π‡πâ‡∏≠‡∏¢‡∏π‡πà

````python
# ‡∏Å‡∏≤‡∏£‡πÉ‡∏ä‡πâ‡∏á‡∏≤‡∏ô Epsilon-Greedy

class EpsilonGreedyBandit:
    """
    ‡∏Ñ‡∏•‡∏≤‡∏™‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÅ‡∏ï‡πà‡∏•‡∏∞ arm (‡πÅ‡∏Ç‡∏ô) ‡∏Ç‡∏≠‡∏á bandit
    
    ‡∏≠‡∏ò‡∏¥‡∏ö‡∏≤‡∏¢:
    - ‡πÅ‡∏ï‡πà‡∏•‡∏∞ arm ‡πÄ‡∏´‡∏°‡∏∑‡∏≠‡∏ô‡∏õ‡∏∏‡πà‡∏°‡πÄ‡∏Ñ‡∏£‡∏∑‡πà‡∏≠‡∏á‡∏™‡∏•‡πá‡∏≠‡∏ï
    - ‡πÄ‡∏£‡∏≤‡πÑ‡∏°‡πà‡∏£‡∏π‡πâ‡∏ß‡πà‡∏≤ arm ‡πÑ‡∏´‡∏ô‡πÉ‡∏´‡πâ‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏°‡∏≤‡∏Å‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î
    - ‡∏ï‡πâ‡∏≠‡∏á‡∏ó‡∏î‡∏•‡∏≠‡∏á‡πÄ‡∏•‡πà‡∏ô‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏õ‡∏£‡∏∞‡∏°‡∏≤‡∏ì‡∏Ñ‡πà‡∏≤
    """
    def __init__(self, p):
        self.p = p              # ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ô‡πà‡∏≤‡∏à‡∏∞‡πÄ‡∏õ‡πá‡∏ô‡∏à‡∏£‡∏¥‡∏á (‡πÄ‡∏£‡∏≤‡πÑ‡∏°‡πà‡∏£‡∏π‡πâ!)
        self.p_estimate = 0.    # ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ô‡πà‡∏≤‡∏à‡∏∞‡πÄ‡∏õ‡πá‡∏ô‡∏ó‡∏µ‡πà‡πÄ‡∏£‡∏≤‡∏õ‡∏£‡∏∞‡∏°‡∏≤‡∏ì
        self.N = 0              # ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏Ñ‡∏£‡∏±‡πâ‡∏á‡∏ó‡∏µ‡πà‡πÄ‡∏£‡∏≤‡∏•‡∏≠‡∏á‡∏î‡∏∂‡∏á arm ‡∏ô‡∏µ‡πâ
    
    def pull(self):
        """
        ‡∏î‡∏∂‡∏á arm ‡πÅ‡∏•‡∏∞‡πÑ‡∏î‡πâ‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏• (0 ‡∏´‡∏£‡∏∑‡∏≠ 1)
        
        ‡∏≠‡∏ò‡∏¥‡∏ö‡∏≤‡∏¢:
        - ‡∏à‡∏≥‡∏•‡∏≠‡∏á‡∏Å‡∏≤‡∏£‡πÑ‡∏î‡πâ‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ï‡∏≤‡∏°‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ô‡πà‡∏≤‡∏à‡∏∞‡πÄ‡∏õ‡πá‡∏ô‡∏à‡∏£‡∏¥‡∏á p
        - ‡∏ñ‡πâ‡∏≤ random() < p ‚Üí ‡πÑ‡∏î‡πâ‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏• (return 1)
        - ‡∏ñ‡πâ‡∏≤ random() >= p ‚Üí ‡πÑ‡∏°‡πà‡πÑ‡∏î‡πâ‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏• (return 0)
        """
        return 1 if np.random.random() < self.p else 0
    
    def update(self, x):
        """
        ‡∏≠‡∏±‡∏û‡πÄ‡∏î‡∏ó‡∏Å‡∏≤‡∏£‡∏õ‡∏£‡∏∞‡∏°‡∏≤‡∏ì‡∏Ñ‡πà‡∏≤‡∏î‡πâ‡∏ß‡∏¢ incremental mean
        
        ‡∏≠‡∏ò‡∏¥‡∏ö‡∏≤‡∏¢:
        - x ‡∏Ñ‡∏∑‡∏≠‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà‡πÑ‡∏î‡πâ‡∏à‡∏≤‡∏Å‡∏Å‡∏≤‡∏£‡∏î‡∏∂‡∏á‡∏Ñ‡∏£‡∏±‡πâ‡∏á‡∏ô‡∏µ‡πâ (0 ‡∏´‡∏£‡∏∑‡∏≠ 1)
        - ‡πÉ‡∏ä‡πâ‡∏™‡∏π‡∏ï‡∏£: mean_new = ((N-1) * mean_old + x) / N
        - ‡πÄ‡∏õ‡πá‡∏ô‡∏Å‡∏≤‡∏£‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏Ñ‡πà‡∏≤‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢‡πÅ‡∏ö‡∏ö‡∏Ñ‡πà‡∏≠‡∏¢ ‡πÜ ‡πÄ‡∏û‡∏¥‡πà‡∏° (‡πÑ‡∏°‡πà‡∏ï‡πâ‡∏≠‡∏á‡πÄ‡∏Å‡πá‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î)
        """
        self.N += 1
        self.p_estimate = ((self.N - 1) * self.p_estimate + x) / self.N

def run_epsilon_greedy(eps, probs, N):
    """
    ‡∏£‡∏±‡∏ô‡∏Å‡∏≤‡∏£‡∏ó‡∏î‡∏•‡∏≠‡∏á Epsilon-Greedy
    
    ‡∏û‡∏≤‡∏£‡∏≤‡∏°‡∏¥‡πÄ‡∏ï‡∏≠‡∏£‡πå:
    - eps: ‡∏Ñ‡πà‡∏≤ epsilon (‡πÄ‡∏ä‡πà‡∏ô 0.1 = ‡∏™‡∏≥‡∏£‡∏ß‡∏à 10%)
    - probs: ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ô‡πà‡∏≤‡∏à‡∏∞‡πÄ‡∏õ‡πá‡∏ô‡∏à‡∏£‡∏¥‡∏á‡∏Ç‡∏≠‡∏á‡πÅ‡∏ï‡πà‡∏•‡∏∞ arm [0.2, 0.5, 0.75]
    - N: ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏Ñ‡∏£‡∏±‡πâ‡∏á‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î‡∏ó‡∏µ‡πà‡∏à‡∏∞‡∏ó‡∏î‡∏•‡∏≠‡∏á
    
    ‡∏Ñ‡∏∑‡∏ô‡∏Ñ‡πà‡∏≤:
    - rewards: ‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà‡πÑ‡∏î‡πâ‡πÉ‡∏ô‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏Ñ‡∏£‡∏±‡πâ‡∏á
    - bandits: ‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£ bandit objects (‡∏°‡∏µ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏™‡∏ñ‡∏¥‡∏ï‡∏¥)
    """
    # ‡∏™‡∏£‡πâ‡∏≤‡∏á bandit ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÅ‡∏ï‡πà‡∏•‡∏∞ arm
    bandits = [EpsilonGreedyBandit(p) for p in probs]
    rewards = np.zeros(N)
    
    for i in range(N):
        # ‡∏ï‡∏±‡∏î‡∏™‡∏¥‡∏ô‡πÉ‡∏à‡πÅ‡∏ö‡∏ö Epsilon-Greedy
        if np.random.random() < eps:
            # ‡∏™‡∏≥‡∏£‡∏ß‡∏à: ‡πÄ‡∏•‡∏∑‡∏≠‡∏Å‡πÅ‡∏ö‡∏ö‡∏™‡∏∏‡πà‡∏°
            j = np.random.randint(len(bandits))
        else:
            # ‡πÉ‡∏ä‡πâ‡∏õ‡∏£‡∏∞‡πÇ‡∏¢‡∏ä‡∏ô‡πå: ‡πÄ‡∏•‡∏∑‡∏≠‡∏Å arm ‡∏ó‡∏µ‡πà‡∏°‡∏µ‡∏Ñ‡πà‡∏≤‡∏õ‡∏£‡∏∞‡∏°‡∏≤‡∏ì‡∏™‡∏π‡∏á‡∏™‡∏∏‡∏î
            j = np.argmax([b.p_estimate for b in bandits])
        
        # ‡∏î‡∏∂‡∏á arm ‡∏ó‡∏µ‡πà‡πÄ‡∏•‡∏∑‡∏≠‡∏Å‡πÅ‡∏•‡∏∞‡πÑ‡∏î‡πâ‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•
        x = bandits[j].pull()
        # ‡∏≠‡∏±‡∏û‡πÄ‡∏î‡∏ó‡∏Å‡∏≤‡∏£‡∏õ‡∏£‡∏∞‡∏°‡∏≤‡∏ì‡∏Ñ‡πà‡∏≤‡∏Ç‡∏≠‡∏á arm ‡∏ô‡∏±‡πâ‡∏ô
        bandits[j].update(x)
        # ‡∏ö‡∏±‡∏ô‡∏ó‡∏∂‡∏Å‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•
        rewards[i] = x
    
    return rewards, bandits

# ========== ‡∏ó‡∏î‡∏™‡∏≠‡∏ö‡∏≠‡∏±‡∏•‡∏Å‡∏≠‡∏£‡∏¥‡∏ó‡∏∂‡∏° ==========

# ‡∏ï‡∏±‡πâ‡∏á‡∏Ñ‡πà‡∏≤‡∏õ‡∏±‡∏ç‡∏´‡∏≤
PROBS = [0.2, 0.5, 0.75]  # ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ô‡πà‡∏≤‡∏à‡∏∞‡πÄ‡∏õ‡πá‡∏ô‡∏à‡∏£‡∏¥‡∏á (‡πÄ‡∏£‡∏≤‡πÑ‡∏°‡πà‡∏£‡∏π‡πâ!)
                           # Arm 1: 20%, Arm 2: 50%, Arm 3: 75% ‚≠ê (‡∏î‡∏µ‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î)
N_TRIALS = 10000           # ‡∏ó‡∏î‡∏•‡∏≠‡∏á 10,000 ‡∏Ñ‡∏£‡∏±‡πâ‡∏á

# ‡∏£‡∏±‡∏ô‡∏Å‡∏≤‡∏£‡∏ó‡∏î‡∏•‡∏≠‡∏á
rewards_eg, bandits_eg = run_epsilon_greedy(eps=0.1, probs=PROBS, N=N_TRIALS)

# ‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå
print("\nüéØ ‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå Epsilon-Greedy (Œµ=0.1):\n")
print("="*60)
for i, b in enumerate(bandits_eg):
    print(f"Arm {i+1}: ‡∏à‡∏£‡∏¥‡∏á={b.p:.2f}, ‡∏õ‡∏£‡∏∞‡∏°‡∏≤‡∏ì={b.p_estimate:.4f}, ‡∏î‡∏∂‡∏á={b.N} ‡∏Ñ‡∏£‡∏±‡πâ‡∏á")

print("="*60)
print(f"\n‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏£‡∏ß‡∏°: {rewards_eg.sum():.0f}")
print(f"‡∏≠‡∏±‡∏ï‡∏£‡∏≤‡∏ä‡∏ô‡∏∞: {rewards_eg.mean():.4f} ({rewards_eg.mean()*100:.2f}%)")
print(f"‡∏Ñ‡πà‡∏≤‡∏ó‡∏µ‡πà‡∏î‡∏µ‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î: {max(PROBS):.2f} ({max(PROBS)*100:.0f}%)")
print(f"‡∏ä‡πà‡∏≠‡∏á‡∏ß‡πà‡∏≤‡∏á: {max(PROBS) - rewards_eg.mean():.4f} ({(max(PROBS) - rewards_eg.mean())/max(PROBS)*100:.1f}% loss)")

print(f"\nüí° ‡∏™‡∏±‡∏á‡πÄ‡∏Å‡∏ï:")
print(f"  ‚úÖ ‡∏≠‡∏±‡∏•‡∏Å‡∏≠‡∏£‡∏¥‡∏ó‡∏∂‡∏°‡πÄ‡∏£‡∏µ‡∏¢‡∏ô‡∏£‡∏π‡πâ‡∏ß‡πà‡∏≤ Arm 3 ‡∏î‡∏µ‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î (‡∏ñ‡∏π‡∏Å‡∏î‡∏∂‡∏á‡∏°‡∏≤‡∏Å‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î)")
print(f"  ‚úÖ ‡∏Ñ‡πà‡∏≤‡∏õ‡∏£‡∏∞‡∏°‡∏≤‡∏ì‡πÉ‡∏Å‡∏•‡πâ‡πÄ‡∏Ñ‡∏µ‡∏¢‡∏á‡∏Ñ‡πà‡∏≤‡∏à‡∏£‡∏¥‡∏á")
print(f"  ‚úÖ ‡πÅ‡∏ï‡πà‡∏¢‡∏±‡∏á‡πÄ‡∏™‡∏µ‡∏¢ {(max(PROBS) - rewards_eg.mean())/max(PROBS)*100:.1f}% ‡πÄ‡∏û‡∏£‡∏≤‡∏∞‡∏ï‡πâ‡∏≠‡∏á‡∏™‡∏≥‡∏£‡∏ß‡∏à")
````

- ‡∏™‡∏∏‡∏î‡∏ó‡πâ‡∏≤‡∏¢: arm ‡∏ó‡∏µ‡πà‡∏î‡∏µ‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î‡∏à‡∏∞‡∏°‡∏µ‡∏Ñ‡πà‡∏≤‡∏õ‡∏£‡∏∞‡∏°‡∏≤‡∏ì‡∏™‡∏π‡∏á‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î ‚Üí ‡∏ñ‡∏π‡∏Å‡πÄ‡∏•‡∏∑‡∏≠‡∏Å‡πÄ‡∏£‡∏∑‡πà‡∏≠‡∏¢ ‡πÜ (exploit)

**‡∏Ç‡πâ‡∏≠‡∏î‡∏µ**: 
- ‚úÖ ‡πÑ‡∏°‡πà‡∏ï‡πâ‡∏≠‡∏á‡∏ï‡∏±‡πâ‡∏á‡∏Ñ‡πà‡∏≤ Œµ

**‡∏Ç‡πâ‡∏≠‡πÄ‡∏™‡∏µ‡∏¢**: 
- ‚ùå ‡∏ï‡πâ‡∏≠‡∏á‡∏£‡∏π‡πâ‡∏ß‡πà‡∏≤‡∏Ñ‡∏ß‡∏£‡∏ï‡∏±‡πâ‡∏á‡∏Ñ‡πà‡∏≤‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏ï‡πâ‡∏ô‡∏ó‡∏µ‡πà‡πÄ‡∏ó‡πà‡∏≤‡πÑ‡∏´‡∏£‡πà
- ‚ùå ‡∏ñ‡πâ‡∏≤‡∏ï‡∏±‡πâ‡∏á‡∏ï‡πà‡∏≥‡πÄ‡∏Å‡∏¥‡∏ô‡πÑ‡∏õ ‡∏à‡∏∞‡πÑ‡∏°‡πà explore ‡∏û‡∏≠
- ‚ùå ‡∏ñ‡πâ‡∏≤‡∏ï‡∏±‡πâ‡∏á‡∏™‡∏π‡∏á‡πÄ‡∏Å‡∏¥‡∏ô‡πÑ‡∏õ ‡∏à‡∏∞ explore ‡∏ô‡∏≤‡∏ô‡πÄ‡∏Å‡∏¥‡∏ô‡πÑ‡∏õ

**‡πÄ‡∏õ‡∏£‡∏µ‡∏¢‡∏ö‡πÄ‡∏ó‡∏µ‡∏¢‡∏ö**: ‡πÄ‡∏´‡∏°‡∏∑‡∏≠‡∏ô‡∏Ñ‡∏ô‡∏ó‡∏µ‡πà‡∏Ñ‡∏¥‡∏î‡∏ß‡πà‡∏≤‡∏ó‡∏∏‡∏Å‡∏£‡πâ‡∏≤‡∏ô‡∏≠‡∏≤‡∏´‡∏≤‡∏£‡πÄ‡∏õ‡πá‡∏ô‡∏£‡πâ‡∏≤‡∏ô‡∏£‡∏∞‡∏î‡∏±‡∏ö 5 ‡∏î‡∏≤‡∏ß (‡∏°‡∏≠‡∏á‡πÇ‡∏•‡∏Å‡πÉ‡∏ô‡πÅ‡∏á‡πà‡∏î‡∏µ)
- ‡πÄ‡∏°‡∏∑‡πà‡∏≠‡∏•‡∏≠‡∏á‡πÅ‡∏•‡πâ‡∏ß‡∏ú‡∏¥‡∏î‡∏´‡∏ß‡∏±‡∏á ‚Üí ‡∏à‡∏∞‡πÑ‡∏õ‡∏•‡∏≠‡∏á‡∏£‡πâ‡∏≤‡∏ô‡∏≠‡∏∑‡πà‡∏ô
- ‡∏™‡∏∏‡∏î‡∏ó‡πâ‡∏≤‡∏¢‡∏à‡∏∞‡∏£‡∏π‡πâ‡∏ß‡πà‡∏≤‡∏£‡πâ‡∏≤‡∏ô‡πÑ‡∏´‡∏ô‡∏î‡∏µ‡∏à‡∏£‡∏¥‡∏á

````python
# ‡∏Å‡∏≤‡∏£‡πÉ‡∏ä‡πâ‡∏á‡∏≤‡∏ô Optimistic Initial Values

class OptimisticBandit:
    """
    Bandit ‡∏ó‡∏µ‡πà‡πÉ‡∏ä‡πâ‡∏Ñ‡πà‡∏≤‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏ï‡πâ‡∏ô‡πÅ‡∏ö‡∏ö‡∏°‡∏≠‡∏á‡πÇ‡∏•‡∏Å‡πÉ‡∏ô‡πÅ‡∏á‡πà‡∏î‡∏µ
    
    ‡∏≠‡∏ò‡∏¥‡∏ö‡∏≤‡∏¢:
    - ‡∏ï‡∏±‡πâ‡∏á‡∏Ñ‡πà‡∏≤‡∏õ‡∏£‡∏∞‡∏°‡∏≤‡∏ì‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏ï‡πâ‡∏ô‡∏™‡∏π‡∏á ‡πÜ (‡πÄ‡∏ä‡πà‡∏ô 5.0 ‡πÅ‡∏ï‡πà‡∏Ñ‡πà‡∏≤‡∏à‡∏£‡∏¥‡∏á‡∏≠‡∏¢‡∏π‡πà‡∏ó‡∏µ‡πà 0-1)
    - ‡∏ó‡∏≥‡πÉ‡∏´‡πâ‡∏≠‡∏±‡∏•‡∏Å‡∏≠‡∏£‡∏¥‡∏ó‡∏∂‡∏°‡∏≠‡∏¢‡∏≤‡∏Å‡∏ó‡∏î‡∏•‡∏≠‡∏á‡∏ó‡∏∏‡∏Å arms (‡πÄ‡∏û‡∏£‡∏≤‡∏∞‡∏Ñ‡∏¥‡∏î‡∏ß‡πà‡∏≤‡∏ó‡∏∏‡∏Å‡∏ï‡∏±‡∏ß‡∏î‡∏µ)
    - ‡πÄ‡∏°‡∏∑‡πà‡∏≠‡∏ó‡∏î‡∏•‡∏≠‡∏á‡πÅ‡∏•‡πâ‡∏ß‡πÑ‡∏î‡πâ‡∏ú‡∏•‡∏ô‡πâ‡∏≠‡∏¢‡∏Å‡∏ß‡πà‡∏≤‡∏ó‡∏µ‡πà‡∏Ñ‡∏¥‡∏î ‚Üí ‡∏Ñ‡πà‡∏≤‡∏õ‡∏£‡∏∞‡∏°‡∏≤‡∏ì‡∏•‡∏î‡∏•‡∏á
    - Arm ‡∏ó‡∏µ‡πà‡∏î‡∏µ‡∏à‡∏£‡∏¥‡∏á‡∏à‡∏∞‡∏°‡∏µ‡∏Ñ‡πà‡∏≤‡∏õ‡∏£‡∏∞‡∏°‡∏≤‡∏ì‡∏™‡∏π‡∏á‡∏Å‡∏ß‡πà‡∏≤‡∏ï‡∏±‡∏ß‡∏≠‡∏∑‡πà‡∏ô
    """
    def __init__(self, p, initial_estimate=5.0):
        self.p = p
        self.p_estimate = initial_estimate  # üåü ‡∏Ñ‡πà‡∏≤‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏ï‡πâ‡∏ô‡∏™‡∏π‡∏á (Optimistic!)
        self.N = 1  # ‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏à‡∏≤‡∏Å 1 (‡∏ó‡∏≥‡πÄ‡∏õ‡πá‡∏ô‡∏ß‡πà‡∏≤‡∏°‡∏µ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• 1 ‡∏ï‡∏±‡∏ß‡πÅ‡∏•‡πâ‡∏ß)
    
    def pull(self):
        """‡∏î‡∏∂‡∏á arm ‡πÅ‡∏•‡∏∞‡πÑ‡∏î‡πâ‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•"""
        return 1 if np.random.random() < self.p else 0
    
    def update(self, x):
        """‡∏≠‡∏±‡∏û‡πÄ‡∏î‡∏ó‡∏Ñ‡πà‡∏≤‡∏õ‡∏£‡∏∞‡∏°‡∏≤‡∏ì"""
        self.N += 1
        self.p_estimate = ((self.N - 1) * self.p_estimate + x) / self.N

def run_optimistic(probs, N, initial=5.0):
    """
    ‡∏£‡∏±‡∏ô‡∏Å‡∏≤‡∏£‡∏ó‡∏î‡∏•‡∏≠‡∏á Optimistic Initial Values
    
    ‡∏û‡∏≤‡∏£‡∏≤‡∏°‡∏¥‡πÄ‡∏ï‡∏≠‡∏£‡πå:
    - probs: ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ô‡πà‡∏≤‡∏à‡∏∞‡πÄ‡∏õ‡πá‡∏ô‡∏à‡∏£‡∏¥‡∏á
    - N: ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏Ñ‡∏£‡∏±‡πâ‡∏á‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î
    - initial: ‡∏Ñ‡πà‡∏≤‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏ï‡πâ‡∏ô (‡∏Ñ‡∏ß‡∏£‡∏™‡∏π‡∏á‡∏Å‡∏ß‡πà‡∏≤‡∏Ñ‡πà‡∏≤‡∏à‡∏£‡∏¥‡∏á)
    
    ‡∏´‡∏°‡∏≤‡∏¢‡πÄ‡∏´‡∏ï‡∏∏:
    - ‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏Å‡∏≤‡∏£‡∏™‡∏∏‡πà‡∏°! ‡πÄ‡∏•‡∏∑‡∏≠‡∏Å arm ‡∏ó‡∏µ‡πà‡∏î‡∏µ‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î‡πÄ‡∏™‡∏°‡∏≠
    - ‡πÅ‡∏ï‡πà‡∏ä‡πà‡∏ß‡∏á‡πÅ‡∏£‡∏Å‡∏à‡∏∞‡∏™‡∏≥‡∏£‡∏ß‡∏à‡∏≠‡∏±‡∏ï‡πÇ‡∏ô‡∏°‡∏±‡∏ï‡∏¥ ‡πÄ‡∏û‡∏£‡∏≤‡∏∞‡∏Ñ‡πà‡∏≤‡∏õ‡∏£‡∏∞‡∏°‡∏≤‡∏ì‡∏¢‡∏±‡∏á‡∏™‡∏π‡∏á‡∏´‡∏°‡∏î
    """
    bandits = [OptimisticBandit(p, initial) for p in probs]
    rewards = np.zeros(N)
    
    for i in range(N):
        # ‡πÄ‡∏•‡∏∑‡∏≠‡∏Å arm ‡∏ó‡∏µ‡πà‡∏°‡∏µ‡∏Ñ‡πà‡∏≤‡∏õ‡∏£‡∏∞‡∏°‡∏≤‡∏ì‡∏™‡∏π‡∏á‡∏™‡∏∏‡∏î‡πÄ‡∏™‡∏°‡∏≠ (‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏Å‡∏≤‡∏£‡∏™‡∏∏‡πà‡∏°!)
        j = np.argmax([b.p_estimate for b in bandits])
        
        x = bandits[j].pull()
        bandits[j].update(x)
        rewards[i] = x
    
    return rewards, bandits

# ========== ‡∏ó‡∏î‡∏™‡∏≠‡∏ö‡∏≠‡∏±‡∏•‡∏Å‡∏≠‡∏£‡∏¥‡∏ó‡∏∂‡∏° ==========

rewards_opt, bandits_opt = run_optimistic(probs=PROBS, N=N_TRIALS, initial=5.0)

print("\nüåü ‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå Optimistic Initial Values (‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏ï‡πâ‡∏ô=5.0):\n")
print("="*60)
for i, b in enumerate(bandits_opt):
    print(f"Arm {i+1}: ‡∏à‡∏£‡∏¥‡∏á={b.p:.2f}, ‡∏õ‡∏£‡∏∞‡∏°‡∏≤‡∏ì={b.p_estimate:.4f}, ‡∏î‡∏∂‡∏á={b.N} ‡∏Ñ‡∏£‡∏±‡πâ‡∏á")

print("="*60)
print(f"\n‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏£‡∏ß‡∏°: {rewards_opt.sum():.0f}")
print(f"‡∏≠‡∏±‡∏ï‡∏£‡∏≤‡∏ä‡∏ô‡∏∞: {rewards_opt.mean():.4f} ({rewards_opt.mean()*100:.2f}%)")

print(f"\nüí° ‡∏™‡∏±‡∏á‡πÄ‡∏Å‡∏ï:")
print(f"  ‚úÖ Arm ‡∏ó‡∏µ‡πà‡∏î‡∏µ‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î (Arm 3) ‡∏ñ‡∏π‡∏Å‡∏î‡∏∂‡∏á‡∏°‡∏≤‡∏Å‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î!")
print(f"  ‚úÖ ‡πÑ‡∏°‡πà‡∏ï‡πâ‡∏≠‡∏á‡∏ï‡∏±‡πâ‡∏á‡∏Ñ‡πà‡∏≤ Œµ (‡∏á‡πà‡∏≤‡∏¢‡∏Å‡∏ß‡πà‡∏≤ Epsilon-Greedy)")
print(f"  ‚úÖ ‡∏™‡∏≥‡∏£‡∏ß‡∏à‡∏≠‡∏±‡∏ï‡πÇ‡∏ô‡∏°‡∏±‡∏ï‡∏¥‡πÉ‡∏ô‡∏ä‡πà‡∏ß‡∏á‡πÅ‡∏£‡∏Å ‡πÅ‡∏•‡πâ‡∏ß‡∏Ñ‡πà‡∏≠‡∏¢ ‡πÜ exploit")
print(f"  ‚úÖ ‡πÄ‡∏ó‡∏Ñ‡∏ô‡∏¥‡∏Ñ‡∏á‡πà‡∏≤‡∏¢ ‡πÜ ‡πÅ‡∏ï‡πà‡πÑ‡∏î‡πâ‡∏ú‡∏•‡∏î‡∏µ!")
````

### üéØ ‡∏≠‡∏±‡∏•‡∏Å‡∏≠‡∏£‡∏¥‡∏ó‡∏∂‡∏°‡∏ó‡∏µ‡πà 3: UCB1 (Upper Confidence Bound - ‡∏Ç‡∏≠‡∏ö‡πÄ‡∏Ç‡∏ï‡∏Ñ‡∏ß‡∏≤‡∏°‡πÄ‡∏ä‡∏∑‡πà‡∏≠‡∏°‡∏±‡πà‡∏ô‡∏î‡πâ‡∏≤‡∏ô‡∏ö‡∏ô)

**‡πÅ‡∏ô‡∏ß‡∏Ñ‡∏¥‡∏î**: 
- ‡πÄ‡∏•‡∏∑‡∏≠‡∏Å arm ‡∏ó‡∏µ‡πà‡∏°‡∏µ **Upper Confidence Bound ‡∏™‡∏π‡∏á‡∏™‡∏∏‡∏î**
- UCB = ‡∏Ñ‡πà‡∏≤‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢ + ‡πÇ‡∏ö‡∏ô‡∏±‡∏™‡∏à‡∏≤‡∏Å‡∏Ñ‡∏ß‡∏≤‡∏°‡πÑ‡∏°‡πà‡πÅ‡∏ô‡πà‡πÉ‡∏à
- Arm ‡∏ó‡∏µ‡πà‡∏•‡∏≠‡∏á‡∏ô‡πâ‡∏≠‡∏¢ ‚Üí ‡πÑ‡∏°‡πà‡πÅ‡∏ô‡πà‡πÉ‡∏à‡∏°‡∏≤‡∏Å ‚Üí ‡πÑ‡∏î‡πâ‡πÇ‡∏ö‡∏ô‡∏±‡∏™‡πÄ‡∏¢‡∏≠‡∏∞ ‚Üí ‡∏°‡∏µ‡πÇ‡∏≠‡∏Å‡∏≤‡∏™‡∏ñ‡∏π‡∏Å‡πÄ‡∏•‡∏∑‡∏≠‡∏Å
- Arm ‡∏ó‡∏µ‡πà‡∏•‡∏≠‡∏á‡πÄ‡∏¢‡∏≠‡∏∞‡πÅ‡∏•‡πâ‡∏ß ‚Üí ‡πÅ‡∏ô‡πà‡πÉ‡∏à ‚Üí ‡πÇ‡∏ö‡∏ô‡∏±‡∏™‡∏ô‡πâ‡∏≠‡∏¢ ‚Üí ‡∏ï‡πâ‡∏≠‡∏á‡∏°‡∏µ‡∏Ñ‡πà‡∏≤‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢‡∏™‡∏π‡∏á‡∏à‡∏£‡∏¥‡∏á ‡πÜ ‡∏ñ‡∏∂‡∏á‡∏à‡∏∞‡∏ñ‡∏π‡∏Å‡πÄ‡∏•‡∏∑‡∏≠‡∏Å

**‡∏™‡∏π‡∏ï‡∏£**:
$$
\text{UCB}(j) = \bar{x}_j + \sqrt{\frac{2 \ln n}{n_j}}
$$

**‡∏™‡∏±‡∏ç‡∏•‡∏±‡∏Å‡∏©‡∏ì‡πå**:
- $\bar{x}_j$ = ‡∏Ñ‡πà‡∏≤‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏Ç‡∏≠‡∏á arm j (‡∏ó‡∏µ‡πà‡πÑ‡∏î‡πâ‡∏à‡∏≤‡∏Å‡∏Å‡∏≤‡∏£‡∏ó‡∏î‡∏•‡∏≠‡∏á)
- $n$ = ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏Ñ‡∏£‡∏±‡πâ‡∏á‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î‡∏ó‡∏µ‡πà‡πÄ‡∏£‡∏≤‡∏ó‡∏î‡∏•‡∏≠‡∏á
- $n_j$ = ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏Ñ‡∏£‡∏±‡πâ‡∏á‡∏ó‡∏µ‡πà‡πÄ‡∏£‡∏≤‡∏ó‡∏î‡∏•‡∏≠‡∏á arm j

**‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á**:
- Arm A: ‡∏Ñ‡πà‡∏≤‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢ = 0.5, ‡∏•‡∏≠‡∏á 100 ‡∏Ñ‡∏£‡∏±‡πâ‡∏á ‚Üí ‡πÇ‡∏ö‡∏ô‡∏±‡∏™ = ‡πÄ‡∏•‡πá‡∏Å ‚Üí UCB ‚âà 0.5 + 0.1 = 0.6
- Arm B: ‡∏Ñ‡πà‡∏≤‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢ = 0.4, ‡∏•‡∏≠‡∏á 10 ‡∏Ñ‡∏£‡∏±‡πâ‡∏á ‚Üí ‡πÇ‡∏ö‡∏ô‡∏±‡∏™ = ‡πÉ‡∏´‡∏ç‡πà ‚Üí UCB ‚âà 0.4 + 0.3 = 0.7 ‚≠ê (‡∏ñ‡∏π‡∏Å‡πÄ‡∏•‡∏∑‡∏≠‡∏Å!)

**‡∏Ç‡πâ‡∏≠‡∏î‡∏µ**: 
- ‚úÖ ‡πÄ‡∏´‡∏°‡∏≤‡∏∞‡∏™‡∏°‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î‡∏ï‡∏≤‡∏°‡∏ó‡∏§‡∏©‡∏é‡∏µ (Optimal regret bound O(log N))
- ‚úÖ ‡πÑ‡∏°‡πà‡∏ï‡πâ‡∏≠‡∏á‡∏ï‡∏±‡πâ‡∏á‡∏û‡∏≤‡∏£‡∏≤‡∏°‡∏¥‡πÄ‡∏ï‡∏≠‡∏£‡πå‡πÉ‡∏î ‡πÜ
- ‚úÖ ‡∏°‡∏µ‡∏û‡∏∑‡πâ‡∏ô‡∏ê‡∏≤‡∏ô‡∏ó‡∏≤‡∏á‡∏Ñ‡∏ì‡∏¥‡∏ï‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå‡∏ó‡∏µ‡πà‡πÅ‡∏Ç‡πá‡∏á‡πÅ‡∏Å‡∏£‡πà‡∏á

**‡∏Ç‡πâ‡∏≠‡πÄ‡∏™‡∏µ‡∏¢**: 
- ‚ùå ‡∏ã‡∏±‡∏ö‡∏ã‡πâ‡∏≠‡∏ô‡∏Å‡∏ß‡πà‡∏≤‡∏≠‡∏±‡∏•‡∏Å‡∏≠‡∏£‡∏¥‡∏ó‡∏∂‡∏°‡∏≠‡∏∑‡πà‡∏ô
- ‚ùå ‡πÉ‡∏ä‡πâ‡πÑ‡∏î‡πâ‡πÄ‡∏â‡∏û‡∏≤‡∏∞‡∏Å‡∏±‡∏ö‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà‡∏°‡∏µ‡∏Ç‡∏≠‡∏ö‡πÄ‡∏Ç‡∏ï (bounded rewards)
- ‚ùå ‡πÑ‡∏°‡πà‡πÑ‡∏î‡πâ‡∏û‡∏¥‡∏à‡∏≤‡∏£‡∏ì‡∏≤‡∏Ñ‡∏ß‡∏≤‡∏°‡πÅ‡∏õ‡∏£‡∏õ‡∏£‡∏ß‡∏ô (variance)

**‡πÄ‡∏õ‡∏£‡∏µ‡∏¢‡∏ö‡πÄ‡∏ó‡∏µ‡∏¢‡∏ö**: ‡πÄ‡∏´‡∏°‡∏∑‡∏≠‡∏ô‡∏Å‡∏≤‡∏£‡πÄ‡∏•‡∏∑‡∏≠‡∏Å‡∏£‡πâ‡∏≤‡∏ô‡∏≠‡∏≤‡∏´‡∏≤‡∏£
- ‡∏£‡πâ‡∏≤‡∏ô‡∏ó‡∏µ‡πà‡∏•‡∏≠‡∏á‡πÅ‡∏•‡πâ‡∏ß 100 ‡∏Ñ‡∏£‡∏±‡πâ‡∏á ‚Üí ‡∏£‡∏π‡πâ‡∏ß‡πà‡∏≤‡∏≠‡∏£‡πà‡∏≠‡∏¢‡πÅ‡∏Ñ‡πà‡πÑ‡∏´‡∏ô‡πÅ‡∏ô‡πà ‡πÜ
- ‡∏£‡πâ‡∏≤‡∏ô‡∏ó‡∏µ‡πà‡∏•‡∏≠‡∏á‡πÅ‡∏Ñ‡πà 5 ‡∏Ñ‡∏£‡∏±‡πâ‡∏á ‚Üí ‡∏¢‡∏±‡∏á‡πÑ‡∏°‡πà‡πÅ‡∏ô‡πà‡πÉ‡∏à ‚Üí ‡∏Ñ‡∏ß‡∏£‡∏•‡∏≠‡∏á‡∏≠‡∏µ‡∏Å‡πÄ‡∏ú‡∏∑‡πà‡∏≠‡∏à‡∏∞‡∏î‡∏µ‡∏Å‡∏ß‡πà‡∏≤‡∏ó‡∏µ‡πà‡∏Ñ‡∏¥‡∏î

````python
# UCB1 Implementation

class UCB1Bandit:
    def __init__(self, p):
        self.p = p
        self.p_estimate = 0.
        self.N = 0
    
    def pull(self):
        return 1 if np.random.random() < self.p else 0
    
    def update(self, x):
        self.N += 1
        self.p_estimate = ((self.N - 1) * self.p_estimate + x) / self.N

def ucb_score(mean, total_pulls, arm_pulls):
    """Calculate UCB1 score"""
    if arm_pulls == 0:
        return float('inf')  # ‡∏¢‡∏±‡∏á‡πÑ‡∏°‡πà‡πÄ‡∏Ñ‡∏¢ pull ‚Üí ‡∏ï‡πâ‡∏≠‡∏á explore
    return mean + np.sqrt(2 * np.log(total_pulls) / arm_pulls)

def run_ucb1(probs, N):
    bandits = [UCB1Bandit(p) for p in probs]
    rewards = np.zeros(N)
    total_pulls = 0
    
    # Initialize: pull each arm once
    for j in range(len(bandits)):
        x = bandits[j].pull()
        bandits[j].update(x)
        total_pulls += 1
    
    for i in range(N):
        # ‡πÄ‡∏•‡∏∑‡∏≠‡∏Å arm ‡∏ó‡∏µ‡πà‡∏°‡∏µ UCB score ‡∏™‡∏π‡∏á‡∏™‡∏∏‡∏î
        j = np.argmax([ucb_score(b.p_estimate, total_pulls, b.N) for b in bandits])
        
        x = bandits[j].pull()
        bandits[j].update(x)
        total_pulls += 1
        rewards[i] = x
    
    return rewards, bandits

# ‡∏ó‡∏î‡∏™‡∏≠‡∏ö
rewards_ucb, bandits_ucb = run_ucb1(probs=PROBS, N=N_TRIALS)

print("\nüéØ UCB1 Results:\n")
for i, b in enumerate(bandits_ucb):
    print(f"Arm {i+1}: true={b.p:.2f}, estimate={b.p_estimate:.4f}, pulls={b.N}")

print(f"\nTotal reward: {rewards_ucb.sum():.0f}")
print(f"Win rate: {rewards_ucb.mean():.4f}")
print(f"\nüí° ‡∏™‡∏±‡∏á‡πÄ‡∏Å‡∏ï: UCB1 ‡πÄ‡∏•‡∏∑‡∏≠‡∏Å optimal arm ‡∏°‡∏≤‡∏Å‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î!")
````

### üéØ Algorithm 4: Thompson Sampling (Bayesian) ‚≠ê

**Idea**: 
- ‡πÉ‡∏ä‡πâ **Beta distribution** ‡πÄ‡∏õ‡πá‡∏ô posterior
- Sample Œ∏ ~ Beta(a, b) ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÅ‡∏ï‡πà‡∏•‡∏∞ arm
- ‡πÄ‡∏•‡∏∑‡∏≠‡∏Å arm ‡∏ó‡∏µ‡πà‡∏°‡∏µ sample ‡∏™‡∏π‡∏á‡∏™‡∏∏‡∏î

**Algorithm**:
```python
for each arm i:
    Œ∏·µ¢ ~ Beta(a·µ¢, b·µ¢)  # Sample

j = argmax(Œ∏·µ¢)      # ‡πÄ‡∏•‡∏∑‡∏≠‡∏Å arm ‡∏ó‡∏µ‡πà‡∏î‡∏µ‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î
```

**Update**:
```python
if reward = 1: a += 1  # Win
if reward = 0: b += 1  # Loss
```

**Pros**: ‚≠ê ‡∏î‡∏µ‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î, Auto-adaptive, No parameters  
**Cons**: ‡∏ã‡∏±‡∏ö‡∏ã‡πâ‡∏≠‡∏ô‡∏´‡∏ô‡πà‡∏≠‡∏¢, ‡∏ï‡πâ‡∏≠‡∏á‡πÄ‡∏Ç‡πâ‡∏≤‡πÉ‡∏à Bayesian

````python
# Thompson Sampling Implementation

class ThompsonBandit:
    def __init__(self, p):
        self.p = p
        self.a = 1  # Beta parameter (wins)
        self.b = 1  # Beta parameter (losses)
        self.N = 0
    
    def pull(self):
        return 1 if np.random.random() < self.p else 0
    
    def sample(self):
        """Sample ‡∏à‡∏≤‡∏Å Beta posterior"""
        return np.random.beta(self.a, self.b)
    
    def update(self, x):
        """Bayesian update"""
        self.a += x
        self.b += 1 - x
        self.N += 1
    
    @property
    def p_estimate(self):
        """Mean of Beta distribution"""
        return self.a / (self.a + self.b)

def run_thompson(probs, N):
    bandits = [ThompsonBandit(p) for p in probs]
    rewards = np.zeros(N)
    
    for i in range(N):
        # Thompson Sampling: Sample ‡πÅ‡∏•‡πâ‡∏ß‡πÄ‡∏•‡∏∑‡∏≠‡∏Å max
        j = np.argmax([b.sample() for b in bandits])
        
        x = bandits[j].pull()
        bandits[j].update(x)
        rewards[i] = x
    
    return rewards, bandits

# ‡∏ó‡∏î‡∏™‡∏≠‡∏ö
rewards_ts, bandits_ts = run_thompson(probs=PROBS, N=N_TRIALS)

print("\n‚≠ê Thompson Sampling Results:\n")
for i, b in enumerate(bandits_ts):
    print(f"Arm {i+1}: true={b.p:.2f}, estimate={b.p_estimate:.4f}, pulls={b.N}, Beta({b.a}, {b.b})")

print(f"\nTotal reward: {rewards_ts.sum():.0f}")
print(f"Win rate: {rewards_ts.mean():.4f}")
print(f"\nüéâ Thompson Sampling: ‡∏î‡∏µ‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î!")
````

### üéØ Use Case: Beyond Binary Outcomes

**‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏ó‡∏µ‡πà‡∏ï‡πâ‡∏≠‡∏á‡πÉ‡∏ä‡πâ Continuous Rewards**:
- üí∞ **Revenue**: ‡∏•‡∏π‡∏Å‡∏Ñ‡πâ‡∏≤‡∏à‡πà‡∏≤‡∏¢‡πÄ‡∏á‡∏¥‡∏ô‡πÄ‡∏ó‡πà‡∏≤‡πÑ‡∏´‡∏£‡πà (0-1000+ ‡∏ö‡∏≤‡∏ó)
- ‚è±Ô∏è **Latency**: ‡∏£‡∏∞‡∏¢‡∏∞‡πÄ‡∏ß‡∏•‡∏≤‡πÉ‡∏ô‡∏Å‡∏≤‡∏£‡πÇ‡∏´‡∏•‡∏î‡∏´‡∏ô‡πâ‡∏≤‡πÄ‡∏ß‡πá‡∏ö (0.1-5 ‡∏ß‡∏¥‡∏ô‡∏≤‡∏ó‡∏µ)
- ‚≠ê **Ratings**: ‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô‡∏£‡∏µ‡∏ß‡∏¥‡∏ß (1-5 ‡∏î‡∏≤‡∏ß)
- üìä **Engagement**: ‡πÄ‡∏ß‡∏•‡∏≤‡∏ó‡∏µ‡πà‡πÉ‡∏ä‡πâ‡∏ö‡∏ô‡∏´‡∏ô‡πâ‡∏≤‡πÄ‡∏ß‡πá‡∏ö (0-600 ‡∏ß‡∏¥‡∏ô‡∏≤‡∏ó‡∏µ)

---

### üßÆ Mathematical Foundation

**Bernoulli Case** (Part 1):
- Reward: r ‚àà {0, 1}
- Prior: Beta(Œ±, Œ≤)
- Likelihood: Bernoulli(p)
- Posterior: Beta(Œ± + wins, Œ≤ + losses)

**Normal Case** (Part 2) ‚≠ê:
- Reward: r ‚àà ‚Ñù (any real number)
- Prior: N(Œº‚ÇÄ, œÉ¬≤‚ÇÄ)
- Likelihood: N(Œº, œÉ¬≤)
- Posterior: N(Œº‚Çô, œÉ¬≤‚Çô)

**Bayesian Update**:
$$
\mu_n = \frac{\tau x + \lambda \mu}{(\tau + \lambda)}
$$
$$
\lambda_n = \lambda + \tau
$$

Where:
- œÑ = precision of likelihood (1/œÉ¬≤)
- Œª = precision of prior (1/œÉ¬≤‚ÇÄ)
- x = observed reward

---

### üí° Key Insight: Normal-Normal Conjugacy

**Why Normal?**
- ‚úÖ Conjugate prior ‚Üí Closed-form posterior (fast!)
- ‚úÖ Central Limit Theorem ‚Üí Many real-world phenomena are normal
- ‚úÖ Mean & Variance naturally capture uncertainty

**Interpretation**:
- Œº (mean) ‚Üí Expected reward
- œÉ¬≤ (variance) ‚Üí Uncertainty about the mean
- As n‚Üë ‚Üí œÉ¬≤‚Üì ‚Üí More certain about mean

---

````python
# Thompson Sampling with Normal Rewards Implementation
# Use Case: Revenue Optimization

class NormalBandit:
    """
    Thompson Sampling ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö continuous rewards (Normal distribution)
    
    Use Cases:
    - Revenue optimization (‡πÄ‡∏á‡∏¥‡∏ô‡∏ó‡∏µ‡πà‡πÑ‡∏î‡πâ‡∏à‡∏≤‡∏Å‡∏•‡∏π‡∏Å‡∏Ñ‡πâ‡∏≤)
    - Latency minimization (‡∏•‡∏î‡πÄ‡∏ß‡∏•‡∏≤‡πÇ‡∏´‡∏•‡∏î)
    - Rating predictions (‡∏ó‡∏≥‡∏ô‡∏≤‡∏¢‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô)
    """
    def __init__(self, true_mean, true_std=1.0):
        self.true_mean = true_mean  # True mean reward (unknown to algorithm)
        self.true_std = true_std    # Assume known variance (common simplification)
        
        # Bayesian parameters for Normal-Normal conjugate
        self.m = 0           # Posterior mean (Œº)
        self.lambda_ = 1     # Posterior precision (1/œÉ¬≤)
        self.tau = 1         # Likelihood precision (1/œÉ¬≤) - assumed known
        self.N = 0           # Number of pulls
    
    def pull(self):
        """Pull arm ‡πÅ‡∏•‡∏∞‡πÑ‡∏î‡πâ reward ‡∏à‡∏≤‡∏Å Normal distribution"""
        return np.random.randn() * self.true_std + self.true_mean
    
    def sample(self):
        """
        Thompson Sampling: Sample Œº ~ N(m, 1/Œª)
        
        Why? ‡πÄ‡∏£‡∏≤‡πÑ‡∏°‡πà‡πÅ‡∏ô‡πà‡πÉ‡∏à‡∏ß‡πà‡∏≤ true mean ‡∏Ñ‡∏∑‡∏≠‡πÄ‡∏ó‡πà‡∏≤‡πÑ‡∏´‡∏£‡πà
        ‚Üí Sample ‡∏à‡∏≤‡∏Å posterior distribution
        ‚Üí Arm ‡∏ó‡∏µ‡πà‡∏°‡∏µ uncertainty ‡∏™‡∏π‡∏á ‚Üí Sample ‡πÑ‡∏î‡πâ‡∏Ñ‡πà‡∏≤‡∏™‡∏π‡∏á/‡∏ï‡πà‡∏≥‡∏°‡∏≤‡∏Å ‚Üí Explore
        ‚Üí Arm ‡∏ó‡∏µ‡πà‡∏°‡∏µ uncertainty ‡∏ï‡πà‡∏≥ ‚Üí Sample ‡πÑ‡∏î‡πâ‡πÉ‡∏Å‡∏•‡πâ mean ‚Üí Exploit
        """
        return np.random.randn() / np.sqrt(self.lambda_) + self.m
    
    def update(self, x):
        """
        Bayesian update ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö Normal-Normal conjugate
        
        Formula:
        Œº‚Çô = (œÑ¬∑x + Œª¬∑Œº) / (œÑ + Œª)
        Œª‚Çô = Œª + œÑ
        
        Interpretation:
        - Œº‚Çô = weighted average ‡∏£‡∏∞‡∏´‡∏ß‡πà‡∏≤‡∏á prior mean ‡∏Å‡∏±‡∏ö observed data
        - Œª‚Çô = precision ‡πÄ‡∏û‡∏¥‡πà‡∏°‡∏Ç‡∏∂‡πâ‡∏ô ‚Üí uncertainty ‡∏•‡∏î‡∏•‡∏á
        """
        self.m = (self.tau * x + self.lambda_ * self.m) / (self.tau + self.lambda_)
        self.lambda_ += self.tau
        self.N += 1
    
    @property
    def posterior_std(self):
        """Standard deviation ‡∏Ç‡∏≠‡∏á posterior"""
        return 1.0 / np.sqrt(self.lambda_)

def run_normal_thompson(true_means, N, true_std=1.0):
    """
    ‡∏£‡∏±‡∏ô Thompson Sampling experiment ‡∏Å‡∏±‡∏ö Normal rewards
    
    Args:
        true_means: List of true mean rewards for each arm
        N: Number of trials
        true_std: True standard deviation (assumed known)
    """
    bandits = [NormalBandit(m, true_std) for m in true_means]
    rewards = np.zeros(N)
    choices = np.zeros(N, dtype=int)
    
    for i in range(N):
        # Thompson Sampling: Sample ‡πÅ‡∏•‡πâ‡∏ß‡πÄ‡∏•‡∏∑‡∏≠‡∏Å max
        j = np.argmax([b.sample() for b in bandits])
        choices[i] = j
        
        # Pull arm ‡πÅ‡∏•‡∏∞‡πÑ‡∏î‡πâ reward (continuous value!)
        x = bandits[j].pull()
        bandits[j].update(x)
        rewards[i] = x
    
    return rewards, bandits, choices

# ‡∏ó‡∏î‡∏™‡∏≠‡∏ö: Revenue Optimization Use Case
print("\nüí∞ Use Case: Revenue Optimization\n")
print("Scenario: ‡πÄ‡∏£‡∏≤‡∏°‡∏µ 3 pricing strategies")
print("  - Strategy A: ‡∏£‡∏≤‡∏Ñ‡∏≤ 100 ‡∏ö‡∏≤‡∏ó ‚Üí avg revenue = 80 ‡∏ö‡∏≤‡∏ó")
print("  - Strategy B: ‡∏£‡∏≤‡∏Ñ‡∏≤ 150 ‡∏ö‡∏≤‡∏ó ‚Üí avg revenue = 120 ‡∏ö‡∏≤‡∏ó")
print("  - Strategy C: ‡∏£‡∏≤‡∏Ñ‡∏≤ 200 ‡∏ö‡∏≤‡∏ó ‚Üí avg revenue = 170 ‡∏ö‡∏≤‡∏ó ‚≠ê")
print("\n‡πÄ‡∏õ‡πâ‡∏≤‡∏´‡∏°‡∏≤‡∏¢: ‡∏´‡∏≤ strategy ‡∏ó‡∏µ‡πà‡πÉ‡∏´‡πâ revenue ‡∏™‡∏π‡∏á‡∏™‡∏∏‡∏î!\n")

# True mean revenues (algorithm ‡πÑ‡∏°‡πà‡∏£‡∏π‡πâ!)
TRUE_MEANS = [80, 120, 170]
TRUE_STD = 20  # Standard deviation ‡∏Ç‡∏≠‡∏á revenue
N_TRIALS = 2000

rewards_normal, bandits_normal, choices_normal = run_normal_thompson(
    true_means=TRUE_MEANS, 
    N=N_TRIALS, 
    true_std=TRUE_STD
)

print("="*70)
print(f"{'Strategy':<15} {'True Mean':<12} {'Estimated Mean':<18} {'Pulls':<10} {'Uncertainty (œÉ)'}")
print("="*70)

for i, b in enumerate(bandits_normal):
    strategy = ['A (‡∏ø100)', 'B (‡∏ø150)', 'C (‡∏ø200) ‚≠ê'][i]
    print(f"{strategy:<15} {b.true_mean:<12.1f} {b.m:<18.2f} {b.N:<10} {b.posterior_std:.4f}")

print("="*70)
print(f"\nTotal Revenue: ‡∏ø{rewards_normal.sum():,.0f}")
print(f"Average Revenue per Customer: ‡∏ø{rewards_normal.mean():.2f}")
print(f"Optimal Strategy (C): ‡∏ø{max(TRUE_MEANS):.0f}")
print(f"Revenue Gap: ‡∏ø{max(TRUE_MEANS) - rewards_normal.mean():.2f} ({(max(TRUE_MEANS) - rewards_normal.mean())/max(TRUE_MEANS)*100:.1f}% loss)")

# Strategy distribution
print(f"\nüìä Strategy Selection Distribution:")
for i in range(len(TRUE_MEANS)):
    count = (choices_normal == i).sum()
    pct = count / N_TRIALS * 100
    bar = '‚ñà' * int(pct / 2)
    print(f"  Strategy {['A','B','C'][i]}: {count:4d} times ({pct:5.1f}%) {bar}")

print(f"\nüí° ‡∏™‡∏±‡∏á‡πÄ‡∏Å‡∏ï:")
print(f"  ‚úÖ Algorithm ‡πÄ‡∏£‡∏µ‡∏¢‡∏ô‡∏£‡∏π‡πâ‡∏ß‡πà‡∏≤ Strategy C ‡∏î‡∏µ‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î!")
print(f"  ‚úÖ ‡∏´‡∏•‡∏±‡∏á‡∏à‡∏≤‡∏Å explore ‡πÅ‡∏•‡πâ‡∏ß ‚Üí ‡πÄ‡∏•‡∏∑‡∏≠‡∏Å C ‡πÄ‡∏Å‡∏∑‡∏≠‡∏ö‡∏ó‡∏∏‡∏Å‡∏Ñ‡∏£‡∏±‡πâ‡∏á")
print(f"  ‚úÖ Uncertainty (œÉ) ‡∏•‡∏î‡∏•‡∏á‡πÄ‡∏£‡∏∑‡πà‡∏≠‡∏¢ ‡πÜ ‚Üí ‡πÅ‡∏ô‡πà‡πÉ‡∏à‡∏°‡∏≤‡∏Å‡∏Ç‡∏∂‡πâ‡∏ô")
````

### üìä Use Case 1: Convergence Speed Analysis

**File**: `convergence.py`

**Question**: ‡∏Ñ‡∏ß‡∏≤‡∏°‡πÅ‡∏ï‡∏Å‡∏ï‡πà‡∏≤‡∏á‡∏Ç‡∏≠‡∏á arms ‡∏°‡∏µ‡∏ú‡∏•‡∏ï‡πà‡∏≠ convergence speed ‡∏´‡∏£‡∏∑‡∏≠‡πÑ‡∏°‡πà?

**Hypothesis**:
- Arms ‡∏ó‡∏µ‡πà **distinct** (‡πÅ‡∏ï‡∏Å‡∏ï‡πà‡∏≤‡∏á‡∏ä‡∏±‡∏î ‡πÄ‡∏ä‡πà‡∏ô 0.1, 0.5, 0.9) ‚Üí Converge ‡πÄ‡∏£‡πá‡∏ß
- Arms ‡∏ó‡∏µ‡πà **similar** (‡πÉ‡∏Å‡∏•‡πâ‡πÄ‡∏Ñ‡∏µ‡∏¢‡∏á ‡πÄ‡∏ä‡πà‡∏ô 0.2, 0.25, 0.3) ‚Üí Converge ‡∏ä‡πâ‡∏≤

**Why?**: 
- Similar arms ‚Üí Hard to distinguish ‚Üí ‡∏ï‡πâ‡∏≠‡∏á sample ‡πÄ‡∏¢‡∏≠‡∏∞
- Distinct arms ‚Üí Easy to distinguish ‚Üí ‡∏£‡∏π‡πâ‡πÄ‡∏£‡πá‡∏ß‡∏ß‡πà‡∏≤‡∏ï‡∏±‡∏ß‡πÑ‡∏´‡∏ô‡∏î‡∏µ

---

````python
# ‡∏Å‡∏≤‡∏£‡∏ß‡∏¥‡πÄ‡∏Ñ‡∏£‡∏≤‡∏∞‡∏´‡πå‡∏Ñ‡∏ß‡∏≤‡∏°‡πÄ‡∏£‡πá‡∏ß‡πÉ‡∏ô‡∏Å‡∏≤‡∏£‡∏•‡∏π‡πà‡πÄ‡∏Ç‡πâ‡∏≤ (Convergence Speed Analysis)
# ‡πÄ‡∏õ‡∏£‡∏µ‡∏¢‡∏ö‡πÄ‡∏ó‡∏µ‡∏¢‡∏ö: Arms ‡∏ó‡∏µ‡πà‡∏Ñ‡∏•‡πâ‡∏≤‡∏¢‡∏Å‡∏±‡∏ô vs Arms ‡∏ó‡∏µ‡πà‡πÅ‡∏ï‡∏Å‡∏ï‡πà‡∏≤‡∏á‡∏Å‡∏±‡∏ô

def analyze_convergence(probs, N, name):
    """
    ‡∏£‡∏±‡∏ô Thompson Sampling ‡πÅ‡∏•‡∏∞‡∏ß‡∏¥‡πÄ‡∏Ñ‡∏£‡∏≤‡∏∞‡∏´‡πå convergence
    
    ‡∏≠‡∏ò‡∏¥‡∏ö‡∏≤‡∏¢:
    - ‡∏£‡∏±‡∏ô‡∏Å‡∏≤‡∏£‡∏ó‡∏î‡∏•‡∏≠‡∏á Thompson Sampling
    - ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏Ñ‡πà‡∏≤‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢‡∏™‡∏∞‡∏™‡∏°
    - ‡∏´‡∏≤‡∏à‡∏∏‡∏î‡∏ó‡∏µ‡πà converge (‡∏ñ‡∏∂‡∏á 95% ‡∏Ç‡∏≠‡∏á‡∏Ñ‡πà‡∏≤‡∏ó‡∏µ‡πà‡∏î‡∏µ‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î)
    """
    # ‡∏£‡∏±‡∏ô Thompson Sampling (‡∏Ñ‡∏∑‡∏ô‡∏Ñ‡πà‡∏≤ 2 ‡∏ï‡∏±‡∏ß: rewards, bandits)
    rewards, bandits = run_thompson(probs, N)
    
    # ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏Ñ‡πà‡∏≤‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢‡∏™‡∏∞‡∏™‡∏°
    cum_avg = np.cumsum(rewards) / (np.arange(N) + 1)
    
    # ‡∏Ñ‡πà‡∏≤‡∏ó‡∏µ‡πà‡∏î‡∏µ‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î (optimal)
    optimal = max(probs)
    
    # ‡∏´‡∏≤ iteration ‡∏ó‡∏µ‡πà converge (‡πÄ‡∏Ç‡πâ‡∏≤‡∏ñ‡∏∂‡∏á 95% ‡∏Ç‡∏≠‡∏á optimal)
    threshold = 0.95 * optimal
    converged_at = np.argmax(cum_avg >= threshold)
    
    return rewards, bandits, cum_avg, converged_at

print("\nüî¨ ‡∏Å‡∏≤‡∏£‡∏ß‡∏¥‡πÄ‡∏Ñ‡∏£‡∏≤‡∏∞‡∏´‡πå‡∏Ñ‡∏ß‡∏≤‡∏°‡πÄ‡∏£‡πá‡∏ß‡πÉ‡∏ô‡∏Å‡∏≤‡∏£‡∏•‡∏π‡πà‡πÄ‡∏Ç‡πâ‡∏≤ (Convergence Speed Analysis)\n")
print("="*70)

# Scenario 1: Similar arms (‡πÉ‡∏Å‡∏•‡πâ‡πÄ‡∏Ñ‡∏µ‡∏¢‡∏á‡∏Å‡∏±‡∏ô)
print("\nüìç ‡∏™‡∏ñ‡∏≤‡∏ô‡∏Å‡∏≤‡∏£‡∏ì‡πå‡∏ó‡∏µ‡πà 1: Arms ‡∏ó‡∏µ‡πà‡∏Ñ‡∏•‡πâ‡∏≤‡∏¢‡∏Å‡∏±‡∏ô (SIMILAR)")
SIMILAR_PROBS = [0.20, 0.25, 0.30]
print(f"Arms: {SIMILAR_PROBS}")
print(f"‡∏ä‡πà‡∏≠‡∏á‡∏ß‡πà‡∏≤‡∏á: max={max(SIMILAR_PROBS):.2f}, min={min(SIMILAR_PROBS):.2f}, ‡∏Ñ‡∏ß‡∏≤‡∏°‡πÅ‡∏ï‡∏Å‡∏ï‡πà‡∏≤‡∏á={max(SIMILAR_PROBS)-min(SIMILAR_PROBS):.2f}")
print("\nüí≠ ‡∏ó‡∏≥‡πÑ‡∏°‡πÄ‡∏£‡∏µ‡∏¢‡∏Å‡∏ß‡πà‡∏≤ '‡∏Ñ‡∏•‡πâ‡∏≤‡∏¢‡∏Å‡∏±‡∏ô'?")
print("   ‚Üí ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ô‡πà‡∏≤‡∏à‡∏∞‡πÄ‡∏õ‡πá‡∏ô‡πÉ‡∏Å‡∏•‡πâ‡πÄ‡∏Ñ‡∏µ‡∏¢‡∏á‡∏Å‡∏±‡∏ô (0.20, 0.25, 0.30)")
print("   ‚Üí ‡∏¢‡∏≤‡∏Å‡∏ó‡∏µ‡πà‡∏à‡∏∞‡πÅ‡∏¢‡∏Å‡πÅ‡∏¢‡∏∞‡∏ß‡πà‡∏≤‡∏ï‡∏±‡∏ß‡πÑ‡∏´‡∏ô‡∏î‡∏µ‡∏Å‡∏ß‡πà‡∏≤‡∏Å‡∏±‡∏ô")
print("   ‚Üí ‡∏ï‡πâ‡∏≠‡∏á‡∏ó‡∏î‡∏•‡∏≠‡∏á‡πÄ‡∏¢‡∏≠‡∏∞‡∏Å‡∏ß‡πà‡∏≤‡∏à‡∏∂‡∏á‡∏à‡∏∞‡πÅ‡∏ô‡πà‡πÉ‡∏à\n")

rewards_sim, bandits_sim, cum_sim, conv_sim = analyze_convergence(SIMILAR_PROBS, N_TRIALS, "Similar")

print(f"‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå:")
for i, b in enumerate(bandits_sim):
    print(f"  Arm {i+1} (p={b.p:.2f}): ‡∏î‡∏∂‡∏á={b.N} ‡∏Ñ‡∏£‡∏±‡πâ‡∏á, ‡∏õ‡∏£‡∏∞‡∏°‡∏≤‡∏ì={b.p_estimate:.4f}")
print(f"‡∏•‡∏π‡πà‡πÄ‡∏Ç‡πâ‡∏≤‡∏ó‡∏µ‡πà: iteration {conv_sim} (~{conv_sim/N_TRIALS*100:.1f}% ‡∏Ç‡∏≠‡∏á‡∏Å‡∏≤‡∏£‡∏ó‡∏î‡∏•‡∏≠‡∏á)")

# Scenario 2: Distinct arms (‡πÅ‡∏ï‡∏Å‡∏ï‡πà‡∏≤‡∏á‡∏ä‡∏±‡∏î)
print("\n" + "="*70)
print("\nüìç ‡∏™‡∏ñ‡∏≤‡∏ô‡∏Å‡∏≤‡∏£‡∏ì‡πå‡∏ó‡∏µ‡πà 2: Arms ‡∏ó‡∏µ‡πà‡πÅ‡∏ï‡∏Å‡∏ï‡πà‡∏≤‡∏á‡∏Å‡∏±‡∏ô (DISTINCT)")
DISTINCT_PROBS = [0.10, 0.50, 0.90]
print(f"Arms: {DISTINCT_PROBS}")
print(f"‡∏ä‡πà‡∏≠‡∏á‡∏ß‡πà‡∏≤‡∏á: max={max(DISTINCT_PROBS):.2f}, min={min(DISTINCT_PROBS):.2f}, ‡∏Ñ‡∏ß‡∏≤‡∏°‡πÅ‡∏ï‡∏Å‡∏ï‡πà‡∏≤‡∏á={max(DISTINCT_PROBS)-min(DISTINCT_PROBS):.2f}")
print("\nüí≠ ‡∏ó‡∏≥‡πÑ‡∏°‡πÄ‡∏£‡∏µ‡∏¢‡∏Å‡∏ß‡πà‡∏≤ '‡πÅ‡∏ï‡∏Å‡∏ï‡πà‡∏≤‡∏á‡∏Å‡∏±‡∏ô'?")
print("   ‚Üí ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ô‡πà‡∏≤‡∏à‡∏∞‡πÄ‡∏õ‡πá‡∏ô‡∏ï‡πà‡∏≤‡∏á‡∏Å‡∏±‡∏ô‡∏°‡∏≤‡∏Å (0.10, 0.50, 0.90)")
print("   ‚Üí ‡πÅ‡∏¢‡∏Å‡πÅ‡∏¢‡∏∞‡∏á‡πà‡∏≤‡∏¢‡∏ß‡πà‡∏≤‡∏ï‡∏±‡∏ß‡πÑ‡∏´‡∏ô‡∏î‡∏µ‡∏Å‡∏ß‡πà‡∏≤‡∏Å‡∏±‡∏ô")
print("   ‚Üí ‡∏ó‡∏î‡∏•‡∏≠‡∏á‡πÑ‡∏°‡πà‡∏Å‡∏µ‡πà‡∏Ñ‡∏£‡∏±‡πâ‡∏á‡∏Å‡πá‡∏£‡∏π‡πâ‡∏ß‡πà‡∏≤‡∏ï‡∏±‡∏ß‡πÑ‡∏´‡∏ô‡∏î‡∏µ‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î\n")

rewards_dist, bandits_dist, cum_dist, conv_dist = analyze_convergence(DISTINCT_PROBS, N_TRIALS, "Distinct")

print(f"‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå:")
for i, b in enumerate(bandits_dist):
    print(f"  Arm {i+1} (p={b.p:.2f}): ‡∏î‡∏∂‡∏á={b.N} ‡∏Ñ‡∏£‡∏±‡πâ‡∏á, ‡∏õ‡∏£‡∏∞‡∏°‡∏≤‡∏ì={b.p_estimate:.4f}")
print(f"‡∏•‡∏π‡πà‡πÄ‡∏Ç‡πâ‡∏≤‡∏ó‡∏µ‡πà: iteration {conv_dist} (~{conv_dist/N_TRIALS*100:.1f}% ‡∏Ç‡∏≠‡∏á‡∏Å‡∏≤‡∏£‡∏ó‡∏î‡∏•‡∏≠‡∏á)")

# ‡πÄ‡∏õ‡∏£‡∏µ‡∏¢‡∏ö‡πÄ‡∏ó‡∏µ‡∏¢‡∏ö‡∏Ñ‡∏ß‡∏≤‡∏°‡πÄ‡∏£‡πá‡∏ß‡πÉ‡∏ô‡∏Å‡∏≤‡∏£‡∏•‡∏π‡πà‡πÄ‡∏Ç‡πâ‡∏≤
print("\n" + "="*70)
print(f"üí° ‡∏Å‡∏≤‡∏£‡πÄ‡∏õ‡∏£‡∏µ‡∏¢‡∏ö‡πÄ‡∏ó‡∏µ‡∏¢‡∏ö Convergence:\n")
print(f"{'‡∏™‡∏ñ‡∏≤‡∏ô‡∏Å‡∏≤‡∏£‡∏ì‡πå':<25} {'‡∏ä‡πà‡∏≠‡∏á‡∏ß‡πà‡∏≤‡∏á':<12} {'‡∏•‡∏π‡πà‡πÄ‡∏Ç‡πâ‡∏≤‡∏ó‡∏µ‡πà':<18} {'‡∏Ñ‡∏ß‡∏≤‡∏°‡πÄ‡∏£‡πá‡∏ß'}")
print("="*70)
print(f"{'Arms ‡∏Ñ‡∏•‡πâ‡∏≤‡∏¢‡∏Å‡∏±‡∏ô':<25} {max(SIMILAR_PROBS)-min(SIMILAR_PROBS):<12.2f} {conv_sim:<18} {'üêå ‡∏ä‡πâ‡∏≤' if conv_sim > conv_dist else 'üöÄ ‡πÄ‡∏£‡πá‡∏ß'}")
print(f"{'Arms ‡πÅ‡∏ï‡∏Å‡∏ï‡πà‡∏≤‡∏á‡∏Å‡∏±‡∏ô':<25} {max(DISTINCT_PROBS)-min(DISTINCT_PROBS):<12.2f} {conv_dist:<18} {'üêå ‡∏ä‡πâ‡∏≤' if conv_dist > conv_sim else 'üöÄ ‡πÄ‡∏£‡πá‡∏ß'}")

speedup = conv_sim / conv_dist if conv_dist > 0 else float('inf')
print(f"\n‡∏Ñ‡∏ß‡∏≤‡∏°‡πÄ‡∏£‡πá‡∏ß‡πÄ‡∏û‡∏¥‡πà‡∏°‡∏Ç‡∏∂‡πâ‡∏ô: {speedup:.1f}x ‡πÄ‡∏£‡πá‡∏ß‡∏Å‡∏ß‡πà‡∏≤‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö arms ‡∏ó‡∏µ‡πà‡πÅ‡∏ï‡∏Å‡∏ï‡πà‡∏≤‡∏á‡∏Å‡∏±‡∏ô!")

# ‡πÅ‡∏™‡∏î‡∏á‡∏Å‡∏£‡∏≤‡∏ü‡πÄ‡∏õ‡∏£‡∏µ‡∏¢‡∏ö‡πÄ‡∏ó‡∏µ‡∏¢‡∏ö
plt.figure(figsize=(14, 6))

# ‡∏Å‡∏£‡∏≤‡∏ü‡∏ã‡πâ‡∏≤‡∏¢: Arms ‡∏ó‡∏µ‡πà‡∏Ñ‡∏•‡πâ‡∏≤‡∏¢‡∏Å‡∏±‡∏ô
plt.subplot(1, 2, 1)
plt.plot(cum_sim, linewidth=2, label='Arms ‡∏Ñ‡∏•‡πâ‡∏≤‡∏¢‡∏Å‡∏±‡∏ô', color='orange')
plt.axhline(max(SIMILAR_PROBS), color='green', linestyle='--', linewidth=2, 
            label=f'‡∏Ñ‡πà‡∏≤‡∏ó‡∏µ‡πà‡∏î‡∏µ‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î ({max(SIMILAR_PROBS):.2f})', alpha=0.7)
plt.axhline(0.95 * max(SIMILAR_PROBS), color='red', linestyle=':', linewidth=1, 
            label='‡πÄ‡∏Å‡∏ì‡∏ë‡πå 95%', alpha=0.5)
plt.axvline(conv_sim, color='red', linestyle='--', alpha=0.5)
plt.xlabel('‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏Ñ‡∏£‡∏±‡πâ‡∏á (Iteration)', fontsize=12)
plt.ylabel('‡∏Ñ‡πà‡∏≤‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏™‡∏∞‡∏™‡∏°', fontsize=12)
plt.title('üêå Arms ‡∏Ñ‡∏•‡πâ‡∏≤‡∏¢‡∏Å‡∏±‡∏ô (‡∏•‡∏π‡πà‡πÄ‡∏Ç‡πâ‡∏≤‡∏ä‡πâ‡∏≤)', fontsize=13, fontweight='bold')
plt.legend(fontsize=10)
plt.grid(alpha=0.3)
plt.xscale('log')

# ‡∏Å‡∏£‡∏≤‡∏ü‡∏Ç‡∏ß‡∏≤: Arms ‡∏ó‡∏µ‡πà‡πÅ‡∏ï‡∏Å‡∏ï‡πà‡∏≤‡∏á‡∏Å‡∏±‡∏ô
plt.subplot(1, 2, 2)
plt.plot(cum_dist, linewidth=2, label='Arms ‡πÅ‡∏ï‡∏Å‡∏ï‡πà‡∏≤‡∏á‡∏Å‡∏±‡∏ô', color='steelblue')
plt.axhline(max(DISTINCT_PROBS), color='green', linestyle='--', linewidth=2, 
            label=f'‡∏Ñ‡πà‡∏≤‡∏ó‡∏µ‡πà‡∏î‡∏µ‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î ({max(DISTINCT_PROBS):.2f})', alpha=0.7)
plt.axhline(0.95 * max(DISTINCT_PROBS), color='red', linestyle=':', linewidth=1, 
            label='‡πÄ‡∏Å‡∏ì‡∏ë‡πå 95%', alpha=0.5)
plt.axvline(conv_dist, color='red', linestyle='--', alpha=0.5)
plt.xlabel('‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏Ñ‡∏£‡∏±‡πâ‡∏á (Iteration)', fontsize=12)
plt.ylabel('‡∏Ñ‡πà‡∏≤‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏™‡∏∞‡∏™‡∏°', fontsize=12)
plt.title('üöÄ Arms ‡πÅ‡∏ï‡∏Å‡∏ï‡πà‡∏≤‡∏á‡∏Å‡∏±‡∏ô (‡∏•‡∏π‡πà‡πÄ‡∏Ç‡πâ‡∏≤‡πÄ‡∏£‡πá‡∏ß)', fontsize=13, fontweight='bold')
plt.legend(fontsize=10)
plt.grid(alpha=0.3)
plt.xscale('log')

plt.tight_layout()
plt.show()

print("\nüéØ ‡∏ö‡∏ó‡πÄ‡∏£‡∏µ‡∏¢‡∏ô‡∏™‡∏≥‡∏Ñ‡∏±‡∏ç:")
print(f"  ‚úÖ Arms ‡πÅ‡∏ï‡∏Å‡∏ï‡πà‡∏≤‡∏á‡∏Å‡∏±‡∏ô ‚Üí ‡πÅ‡∏¢‡∏Å‡πÅ‡∏¢‡∏∞‡∏á‡πà‡∏≤‡∏¢ ‚Üí ‡∏•‡∏π‡πà‡πÄ‡∏Ç‡πâ‡∏≤‡πÄ‡∏£‡πá‡∏ß üöÄ")
print(f"  ‚úÖ Arms ‡∏Ñ‡∏•‡πâ‡∏≤‡∏¢‡∏Å‡∏±‡∏ô ‚Üí ‡πÅ‡∏¢‡∏Å‡πÅ‡∏¢‡∏∞‡∏¢‡∏≤‡∏Å ‚Üí ‡∏•‡∏π‡πà‡πÄ‡∏Ç‡πâ‡∏≤‡∏ä‡πâ‡∏≤ üêå")
print(f"  ‚úÖ ‡∏Å‡∏≤‡∏£‡πÉ‡∏ä‡πâ‡∏á‡∏≤‡∏ô‡∏à‡∏£‡∏¥‡∏á: ‡∏ñ‡πâ‡∏≤‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏∑‡∏≠‡∏Å‡∏Ñ‡∏•‡πâ‡∏≤‡∏¢‡∏Å‡∏±‡∏ô‡πÄ‡∏Å‡∏¥‡∏ô‡πÑ‡∏õ ‡∏Ñ‡∏ß‡∏£‡∏õ‡∏£‡∏±‡∏ö‡πÉ‡∏´‡πâ‡πÅ‡∏ï‡∏Å‡∏ï‡πà‡∏≤‡∏á‡∏ä‡∏±‡∏î!")
print(f"  ")
print(f"  üì± ‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏ó‡∏µ‡πà‡πÑ‡∏°‡πà‡∏î‡∏µ: ‡∏ó‡∏î‡∏™‡∏≠‡∏ö‡∏õ‡∏∏‡πà‡∏° '‡∏ã‡∏∑‡πâ‡∏≠‡πÄ‡∏•‡∏¢' vs '‡∏ã‡∏∑‡πâ‡∏≠‡∏ó‡∏±‡∏ô‡∏ó‡∏µ' (‡∏Ñ‡∏•‡πâ‡∏≤‡∏¢‡∏Å‡∏±‡∏ô‡πÄ‡∏Å‡∏¥‡∏ô‡πÑ‡∏õ)")
print(f"  üì± ‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏ó‡∏µ‡πà‡∏î‡∏µ: ‡∏ó‡∏î‡∏™‡∏≠‡∏ö‡∏õ‡∏∏‡πà‡∏° '‡∏ã‡∏∑‡πâ‡∏≠‡πÄ‡∏•‡∏¢' vs '‡πÉ‡∏™‡πà‡∏ï‡∏∞‡∏Å‡∏£‡πâ‡∏≤' (‡∏Å‡∏£‡∏∞‡∏ö‡∏ß‡∏ô‡∏Å‡∏≤‡∏£‡∏ï‡πà‡∏≤‡∏á‡∏Å‡∏±‡∏ô‡∏ä‡∏±‡∏î)")
print(f"  ")
print(f"  üé® ‡∏™‡∏µ: '‡πÅ‡∏î‡∏á #FF0000' vs '‡πÅ‡∏î‡∏á #FF0011' (‡∏ï‡πà‡∏≤‡∏á‡∏Å‡∏±‡∏ô‡πÄ‡∏•‡πá‡∏Å‡∏ô‡πâ‡∏≠‡∏¢ = ‡∏ä‡πâ‡∏≤)")
print(f"  üé® ‡∏™‡∏µ: '‡πÅ‡∏î‡∏á #FF0000' vs '‡∏ü‡πâ‡∏≤ #0000FF' (‡∏ï‡πà‡∏≤‡∏á‡∏Å‡∏±‡∏ô‡∏ä‡∏±‡∏î = ‡πÄ‡∏£‡πá‡∏ß)")
````

### üìä Use Case 2: Epsilon Parameter Tuning

**File**: `comparing_epsilons.py`

**Question**: ‡∏Ñ‡πà‡∏≤ Œµ ‡∏ó‡∏µ‡πà‡πÄ‡∏´‡∏°‡∏≤‡∏∞‡∏™‡∏°‡∏Ñ‡∏∑‡∏≠‡πÄ‡∏ó‡πà‡∏≤‡πÑ‡∏´‡∏£‡πà?

**Candidates**:
- Œµ = 0.10 ‚Üí Explore 10% of the time
- Œµ = 0.05 ‚Üí Explore 5% of the time  
- Œµ = 0.01 ‚Üí Explore 1% of the time

**Trade-off**:
- Œµ ‡∏™‡∏π‡∏á ‚Üí Explore ‡∏°‡∏≤‡∏Å ‚Üí ‡πÄ‡∏£‡∏µ‡∏¢‡∏ô‡∏£‡∏π‡πâ‡πÄ‡∏£‡πá‡∏ß ‡πÅ‡∏ï‡πà waste reward
- Œµ ‡∏ï‡πà‡∏≥ ‚Üí Exploit ‡∏°‡∏≤‡∏Å ‚Üí ‡πÑ‡∏î‡πâ reward ‡πÄ‡∏£‡πá‡∏ß ‡πÅ‡∏ï‡πà‡∏≠‡∏≤‡∏à stuck ‡∏ó‡∏µ‡πà suboptimal

**Goal**: ‡∏´‡∏≤ Œµ ‡∏ó‡∏µ‡πà‡πÉ‡∏´‡πâ total reward ‡∏™‡∏π‡∏á‡∏™‡∏∏‡∏î!

---

````python
# Epsilon Parameter Tuning
# Compare Œµ = 0.1, 0.05, 0.01

print("\nüîß Epsilon Parameter Tuning\n")
print("="*70)
print(f"Testing Œµ values: [0.1, 0.05, 0.01]")
print(f"Arms: {PROBS}")
print(f"Trials: {N_TRIALS}\n")

epsilons = [0.1, 0.05, 0.01]
results_eps = {}

for eps in epsilons:
    print(f"Running Œµ = {eps}...")
    rewards, bandits = run_epsilon_greedy(eps=eps, probs=PROBS, N=N_TRIALS)
    results_eps[eps] = {
        'rewards': rewards,
        'bandits': bandits,
        'cum_avg': np.cumsum(rewards) / (np.arange(N_TRIALS) + 1),
        'total': rewards.sum(),
        'final_mean': rewards.mean()
    }

print("\n" + "="*70)
print(f"{'Epsilon (Œµ)':<15} {'Total Reward':<15} {'Avg Reward':<15} {'Gap from Optimal'}")
print("="*70)

optimal = max(PROBS)
for eps in epsilons:
    r = results_eps[eps]
    gap = optimal - r['final_mean']
    print(f"{eps:<15.2f} {r['total']:<15.0f} {r['final_mean']:<15.4f} {gap:.4f} ({gap/optimal*100:.2f}%)")

print(f"\nOptimal: {optimal:.4f}")

# Find best epsilon
best_eps = max(epsilons, key=lambda e: results_eps[e]['total'])
print(f"\n‚≠ê Best Œµ: {best_eps} (highest total reward)")

# Visualize comparison
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Plot 1: Cumulative Average (Log scale)
ax1 = axes[0, 0]
colors = ['red', 'orange', 'blue']
for eps, color in zip(epsilons, colors):
    ax1.plot(results_eps[eps]['cum_avg'], label=f'Œµ = {eps}', 
            linewidth=2, alpha=0.8, color=color)

ax1.axhline(optimal, color='green', linestyle='--', linewidth=2, 
           label=f'Optimal ({optimal:.2f})', alpha=0.7)
ax1.set_xlabel('Iteration', fontsize=12)
ax1.set_ylabel('Cumulative Average Reward', fontsize=12)
ax1.set_title('üìà Convergence Comparison (Log Scale)', fontsize=13, fontweight='bold')
ax1.legend(fontsize=11)
ax1.grid(alpha=0.3)
ax1.set_xscale('log')

# Plot 2: Cumulative Average (Linear scale)
ax2 = axes[0, 1]
for eps, color in zip(epsilons, colors):
    ax2.plot(results_eps[eps]['cum_avg'], label=f'Œµ = {eps}', 
            linewidth=2, alpha=0.8, color=color)

ax2.axhline(optimal, color='green', linestyle='--', linewidth=2, 
           label=f'Optimal ({optimal:.2f})', alpha=0.7)
ax2.set_xlabel('Iteration', fontsize=12)
ax2.set_ylabel('Cumulative Average Reward', fontsize=12)
ax2.set_title('üìà Convergence Comparison (Linear Scale)', fontsize=13, fontweight='bold')
ax2.legend(fontsize=11)
ax2.grid(alpha=0.3)

# Plot 3: Cumulative Regret
ax3 = axes[1, 0]
for eps, color in zip(epsilons, colors):
    optimal_rewards = np.ones(N_TRIALS) * optimal
    regret = np.cumsum(optimal_rewards - results_eps[eps]['rewards'])
    ax3.plot(regret, label=f'Œµ = {eps} (final: {regret[-1]:.0f})', 
            linewidth=2, alpha=0.8, color=color)

ax3.set_xlabel('Iteration', fontsize=12)
ax3.set_ylabel('Cumulative Regret', fontsize=12)
ax3.set_title('üìâ Cumulative Regret Comparison', fontsize=13, fontweight='bold')
ax3.legend(fontsize=11)
ax3.grid(alpha=0.3)

# Plot 4: Final Performance Bar Chart
ax4 = axes[1, 1]
total_rewards = [results_eps[eps]['total'] for eps in epsilons]
bars = ax4.bar([f'Œµ={e}' for e in epsilons], total_rewards, 
               color=colors, alpha=0.7, edgecolor='black', linewidth=2)

# Add value labels on bars
for bar in bars:
    height = bar.get_height()
    ax4.text(bar.get_x() + bar.get_width()/2., height,
            f'{height:.0f}',
            ha='center', va='bottom', fontsize=12, fontweight='bold')

# Optimal line
ax4.axhline(optimal * N_TRIALS, color='green', linestyle='--', 
           linewidth=2, label=f'Optimal ({optimal * N_TRIALS:.0f})', alpha=0.7)

ax4.set_ylabel('Total Reward', fontsize=12)
ax4.set_title('üèÜ Total Reward Comparison', fontsize=13, fontweight='bold')
ax4.legend(fontsize=11)
ax4.grid(alpha=0.3, axis='y')

# Highlight best
best_idx = epsilons.index(best_eps)
bars[best_idx].set_edgecolor('gold')
bars[best_idx].set_linewidth(4)

plt.tight_layout()
plt.show()

print("\nüí° Key Insights:\n")
print(f"1. Œµ = 0.10 (High Exploration):")
print(f"   ‚úÖ Learns fastest (converges early)")
print(f"   ‚ùå Wastes 10% of trials on suboptimal arms")
print(f"   ‚Üí Good for SHORT experiments")

print(f"\n2. Œµ = 0.05 (Medium Exploration):")
print(f"   ‚úÖ Good balance between exploration & exploitation")
print(f"   ‚úÖ Usually BEST for medium-length experiments")
print(f"   ‚Üí ‚≠ê RECOMMENDED for most use cases")

print(f"\n3. Œµ = 0.01 (Low Exploration):")
print(f"   ‚úÖ Exploits heavily ‚Üí High reward if converged")
print(f"   ‚ùå May converge slowly or get stuck")
print(f"   ‚Üí Good for LONG experiments (N > 10,000)")

print(f"\nüéØ General Rule:")
print(f"   - Short experiments (N < 1000): Œµ = 0.10")
print(f"   - Medium experiments (N ~ 10,000): Œµ = 0.05 ‚≠ê")
print(f"   - Long experiments (N > 100,000): Œµ = 0.01")
print(f"   - OR: Use Thompson Sampling (no parameter tuning needed!) üéâ")
````

## üìä Part 3: Statistical Testing

### üî¨ Chi-Square Test

**‡πÉ‡∏ä‡πâ‡πÄ‡∏°‡∏∑‡πà‡∏≠‡πÑ‡∏£**: ‡πÄ‡∏õ‡∏£‡∏µ‡∏¢‡∏ö‡πÄ‡∏ó‡∏µ‡∏¢‡∏ö categorical data (click/no-click)

**Contingency Table**:
```
        Click    No Click
Ad A    a        b
Ad B    c        d
```

**Chi-Square Statistic**:
$$
\chi^2 = \frac{(ad - bc)^2 (a+b+c+d)}{(a+b)(c+d)(a+c)(b+d)}
$$

**H‚ÇÄ**: Ad A ‡πÅ‡∏•‡∏∞ Ad B ‡∏°‡∏µ click rate ‡πÄ‡∏ó‡πà‡∏≤‡∏Å‡∏±‡∏ô  
**H‚ÇÅ**: Ad A ‡πÅ‡∏•‡∏∞ Ad B ‡∏°‡∏µ click rate ‡πÑ‡∏°‡πà‡πÄ‡∏ó‡πà‡∏≤‡∏Å‡∏±‡∏ô

````python
# Chi-Square Test Demo

from scipy.stats import chi2, chi2_contingency

def chi_square_test(p1, p2, N):
    """
    ‡∏£‡∏±‡∏ô Chi-Square test ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö A/B testing
    """
    # Generate data
    clicks_A = np.random.binomial(N, p1)
    clicks_B = np.random.binomial(N, p2)
    
    # Contingency table
    table = np.array([
        [clicks_A, N - clicks_A],      # Ad A
        [clicks_B, N - clicks_B]       # Ad B
    ])
    
    # Chi-square test
    chi2_stat, p_value, dof, expected = chi2_contingency(table, correction=False)
    
    return table, chi2_stat, p_value

# ‡∏ó‡∏î‡∏™‡∏≠‡∏ö
print("\nüî¨ Chi-Square Test Demo\n")

# Case 1: Similar probabilities (p1 ‚âà p2)
table1, chi2_1, p1_val = chi_square_test(p1=0.10, p2=0.11, N=1000)
print("Case 1: Similar probabilities (p1=0.10, p2=0.11)")
print(f"Contingency Table:\n{table1}")
print(f"œá¬≤ statistic: {chi2_1:.4f}")
print(f"p-value: {p1_val:.4f}")
print(f"Conclusion: {'‚ùå No significant difference' if p1_val > 0.05 else '‚úÖ Significant difference'}")

print("\n" + "="*50 + "\n")

# Case 2: Different probabilities (p1 << p2)
table2, chi2_2, p2_val = chi_square_test(p1=0.10, p2=0.20, N=1000)
print("Case 2: Different probabilities (p1=0.10, p2=0.20)")
print(f"Contingency Table:\n{table2}")
print(f"œá¬≤ statistic: {chi2_2:.4f}")
print(f"p-value: {p2_val:.4f}")
print(f"Conclusion: {'‚ùå No significant difference' if p2_val > 0.05 else '‚úÖ Significant difference'}")

print("\nüí° ‡∏™‡∏±‡∏á‡πÄ‡∏Å‡∏ï:")
print("  - p-value < 0.05 ‚Üí Reject H‚ÇÄ ‚Üí ‡∏°‡∏µ‡∏Ñ‡∏ß‡∏≤‡∏°‡πÅ‡∏ï‡∏Å‡∏ï‡πà‡∏≤‡∏á‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏°‡∏µ‡∏ô‡∏±‡∏¢‡∏™‡∏≥‡∏Ñ‡∏±‡∏ç")
print("  - p-value > 0.05 ‚Üí Fail to reject H‚ÇÄ ‚Üí ‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏Ñ‡∏ß‡∏≤‡∏°‡πÅ‡∏ï‡∏Å‡∏ï‡πà‡∏≤‡∏á")
````

### üî¨ T-Test

**‡πÉ‡∏ä‡πâ‡πÄ‡∏°‡∏∑‡πà‡∏≠‡πÑ‡∏£**: ‡πÄ‡∏õ‡∏£‡∏µ‡∏¢‡∏ö‡πÄ‡∏ó‡∏µ‡∏¢‡∏ö means ‡∏Ç‡∏≠‡∏á 2 ‡∏Å‡∏•‡∏∏‡πà‡∏° (continuous data)

**T-Statistic**:
$$
t = \frac{\bar{x}_1 - \bar{x}_2}{s \sqrt{\frac{2}{n}}}
$$

Where:
- $\bar{x}_1, \bar{x}_2$ = means ‡∏Ç‡∏≠‡∏á‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏Å‡∏•‡∏∏‡πà‡∏°
- $s$ = pooled standard deviation
- $n$ = sample size

**H‚ÇÄ**: Œº‚ÇÅ = Œº‚ÇÇ (means ‡πÄ‡∏ó‡πà‡∏≤‡∏Å‡∏±‡∏ô)  
**H‚ÇÅ**: Œº‚ÇÅ ‚â† Œº‚ÇÇ (means ‡πÑ‡∏°‡πà‡πÄ‡∏ó‡πà‡∏≤‡∏Å‡∏±‡∏ô)

````python
# T-Test Demo

from scipy.stats import ttest_ind

print("\nüî¨ T-Test Demo\n")

# Generate data
N = 100
group_A = np.random.randn(N) + 2  # mean = 2, std = 1
group_B = np.random.randn(N) + 2.5  # mean = 2.5, std = 1

# T-test
t_stat, p_value = ttest_ind(group_A, group_B)

print(f"Group A: mean={group_A.mean():.4f}, std={group_A.std(ddof=1):.4f}")
print(f"Group B: mean={group_B.mean():.4f}, std={group_B.std(ddof=1):.4f}")
print(f"\nt-statistic: {t_stat:.4f}")
print(f"p-value: {p_value:.4f}")
print(f"\nConclusion: {'‚úÖ Significant difference' if p_value < 0.05 else '‚ùå No significant difference'}")

# Visualize
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(group_A, bins=20, alpha=0.6, label='Group A', color='blue')
plt.hist(group_B, bins=20, alpha=0.6, label='Group B', color='red')
plt.xlabel('Value', fontsize=11)
plt.ylabel('Frequency', fontsize=11)
plt.title('Distribution Comparison', fontsize=12)
plt.legend()
plt.grid(alpha=0.3)

plt.subplot(1, 2, 2)
plt.boxplot([group_A, group_B], labels=['Group A', 'Group B'])
plt.ylabel('Value', fontsize=11)
plt.title('Box Plot Comparison', fontsize=12)
plt.grid(alpha=0.3, axis='y')

plt.tight_layout()
plt.show()
````

## üé® Part 4: Bayesian Learning Visualization

### üìö Use Case: Understanding How Bayesian Learning Works

**From** `demo.py`: Visualize Beta distribution evolution

**Key Concept**: 
- **Prior** ‚Üí **Likelihood** ‚Üí **Posterior**
- As data arrives ‚Üí Distribution concentrates around true value
- Uncertainty decreases over time

**Why Important?**:
- üí° Build intuition for Bayesian inference
- üéì Teaching tool for students
- üî¨ Understand exploration vs exploitation visually

---

````python
# Beta Distribution Evolution Visualization
# Replicate demo.py: Show how uncertainty decreases with data

from scipy.stats import beta as beta_dist

def plot_beta_evolution(true_ctr, snapshots=[0, 5, 10, 25, 50, 100, 500, 1000]):
    """
    ‡πÅ‡∏™‡∏î‡∏á‡∏ß‡∏¥‡∏ß‡∏±‡∏í‡∏ô‡∏≤‡∏Å‡∏≤‡∏£‡∏Ç‡∏≠‡∏á Beta distribution ‡πÄ‡∏°‡∏∑‡πà‡∏≠‡πÑ‡∏î‡πâ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏°‡∏≤‡∏Å‡∏Ç‡∏∂‡πâ‡∏ô
    
    Args:
        true_ctr: True click-through rate (probability)
        snapshots: Iterations to visualize
    """
    # Initialize Beta(1,1) = Uniform prior
    a, b = 1, 1
    
    # Setup subplots
    n_plots = len(snapshots)
    n_cols = 4
    n_rows = (n_plots + n_cols - 1) // n_cols
    
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, 4*n_rows))
    axes = axes.flatten() if n_rows > 1 else [axes] if n_cols == 1 else axes
    
    x = np.linspace(0, 1, 200)
    
    snapshot_idx = 0
    
    for t in range(max(snapshots) + 1):
        # Simulate coin toss
        if t > 0:  # Skip first iteration (prior only)
            result = (np.random.random() < true_ctr)
            if result:
                a += 1
            else:
                b += 1
        
        # Plot at snapshot points
        if t in snapshots:
            ax = axes[snapshot_idx]
            
            # Calculate Beta distribution
            y = beta_dist.pdf(x, a, b)
            mean = a / (a + b)
            mode = (a - 1) / (a + b - 2) if a > 1 and b > 1 else None
            
            # Plot distribution
            ax.plot(x, y, linewidth=2, color='steelblue')
            ax.fill_between(x, 0, y, alpha=0.3, color='steelblue')
            
            # Mark true CTR
            ax.axvline(true_ctr, color='red', linestyle='--', linewidth=2, 
                      label=f'True CTR = {true_ctr:.2f}', alpha=0.7)
            
            # Mark mean
            ax.axvline(mean, color='green', linestyle=':', linewidth=2, 
                      label=f'Mean = {mean:.2f}', alpha=0.7)
            
            # Title and labels
            ax.set_title(f'After {t} trials: Beta({a}, {b})', fontsize=11, fontweight='bold')
            ax.set_xlabel('Probability (p)', fontsize=10)
            ax.set_ylabel('Density', fontsize=10)
            ax.legend(fontsize=9)
            ax.grid(alpha=0.3)
            ax.set_ylim(0, max(y)*1.1)
            
            # Add text box with stats
            stats_text = f'Clicks: {a-1}\nNo-Clicks: {b-1}'
            ax.text(0.02, 0.98, stats_text, transform=ax.transAxes,
                   fontsize=9, verticalalignment='top',
                   bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))
            
            snapshot_idx += 1
    
    # Hide unused subplots
    for idx in range(snapshot_idx, len(axes)):
        axes[idx].axis('off')
    
    plt.tight_layout()
    plt.show()

# Run visualization
print("\nüé® Bayesian Learning: From Uncertainty to Certainty\n")
print("Scenario: ‡πÄ‡∏£‡∏≤‡∏°‡∏µ advertisement ‡∏ó‡∏µ‡πà‡∏°‡∏µ true CTR = 30%")
print("Question: ‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏à‡∏≤‡∏Å‡πÑ‡∏°‡πà‡∏£‡∏π‡πâ‡∏≠‡∏∞‡πÑ‡∏£‡πÄ‡∏•‡∏¢ ‚Üí ‡∏à‡∏∞‡πÄ‡∏£‡∏µ‡∏¢‡∏ô‡∏£‡∏π‡πâ‡πÑ‡∏î‡πâ‡∏≠‡∏¢‡πà‡∏≤‡∏á‡πÑ‡∏£?\n")
print("Answer: Bayesian Learning! üìä\n")

np.random.seed(42)
plot_beta_evolution(true_ctr=0.3, snapshots=[0, 5, 10, 25, 50, 100, 500, 1000])

print("\nüí° Key Observations:\n")
print("1. üìç After 0 trials (Prior):")
print("   - Beta(1,1) = Uniform[0,1] ‚Üí No knowledge!")
print("   - All probabilities equally likely")
print("   - Maximum uncertainty")

print("\n2. üìç After 5-10 trials:")
print("   - Distribution starts to concentrate")
print("   - Still quite wide ‚Üí Still uncertain")
print("   - Mean moves toward true value")

print("\n3. üìç After 25-50 trials:")
print("   - Distribution narrower")
print("   - Mean closer to true CTR (30%)")
print("   - Reduced uncertainty")

print("\n4. üìç After 100-1000 trials:")
print("   - Very narrow distribution")
print("   - Mean ‚âà true CTR")
print("   - High confidence!")

print("\nüéØ This is WHY Thompson Sampling works:")
print("   ‚úÖ Early: Wide distribution ‚Üí Sample varies ‚Üí Explore")
print("   ‚úÖ Late: Narrow distribution ‚Üí Sample stable ‚Üí Exploit")
print("   ‚úÖ Automatic balance: No need to set Œµ!")
````

### üìù API Specification

#### **Endpoint 1**: GET /get_ad

**Purpose**: ‡πÄ‡∏•‡∏∑‡∏≠‡∏Å ad ‡∏ó‡∏µ‡πà‡∏î‡∏µ‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î‡∏î‡πâ‡∏ß‡∏¢ Thompson Sampling

**Algorithm**:
```python
Œ∏_A ~ Beta(a_A, b_A)  # Sample ‡∏à‡∏≤‡∏Å posterior A
Œ∏_B ~ Beta(a_B, b_B)  # Sample ‡∏à‡∏≤‡∏Å posterior B

if Œ∏_A > Œ∏_B:
    return {"advertisement_id": "A"}
else:
    return {"advertisement_id": "B"}
```

**Response**:
```json
{
    "advertisement_id": "A"  // or "B"
}
```

---

#### **Endpoint 2**: POST /click_ad

**Purpose**: Update statistics ‡πÄ‡∏°‡∏∑‡πà‡∏≠ user ‡∏Ñ‡∏•‡∏¥‡∏Å ad

**Request Body**:
```json
{
    "advertisement_id": "A"  // Which ad was clicked
}
```

**Bayesian Update**:
```python
if clicked:
    bandit.clicks += 1     # a += 1 (success)
    
# Always count view:
bandit.views += 1          # Total trials
# b = views - clicks (failures)
```

**Response**:
```json
{
    "result": "OK"
}
```

---

### üíª Server Implementation (Simplified)

**Key Components**:

1. **Bandit Class**:
   - Tracks `clicks` ‡πÅ‡∏•‡∏∞ `views`
   - Method `sample()`: Sample Œ∏ ~ Beta(1+clicks, 1+views-clicks)
   - Methods `add_click()`, `add_view()`: Update statistics

2. **Flask Routes**:
   - `/get_ad`: Thompson Sampling selection
   - `/click_ad`: Bayesian update

**Why Flask?**:
- ‚úÖ Lightweight Python web framework
- ‚úÖ Easy to deploy
- ‚úÖ RESTful API pattern
- ‚úÖ Stateless (can use database for state)

````python
# Simulated Client-Server Demo (Without HTTP)
# ‡πÅ‡∏™‡∏î‡∏á logic ‡∏Ç‡∏≠‡∏á server_solution.py + client.py ‡πÅ‡∏ö‡∏ö in-memory

class ProductionBandit:
    """
    Bandit class ‡∏ï‡∏≤‡∏° server_solution.py
    ‡πÉ‡∏ä‡πâ‡πÉ‡∏ô production API
    """
    def __init__(self, name):
        self.name = name
        self.clicks = 0
        self.views = 0
    
    def sample(self):
        """Thompson Sampling: Sample ‡∏à‡∏≤‡∏Å Beta posterior"""
        a = 1 + self.clicks
        b = 1 + self.views - self.clicks
        return np.random.beta(a, b)
    
    def add_click(self):
        """User ‡∏Ñ‡∏•‡∏¥‡∏Å ad"""
        self.clicks += 1
    
    def add_view(self):
        """User ‡πÄ‡∏´‡πá‡∏ô ad"""
        self.views += 1
        
        # Log every 50 views (as in server_solution.py)
        if self.views % 50 == 0:
            ctr = self.clicks / self.views if self.views > 0 else 0
            print(f"  üìä {self.name}: views={self.views}, clicks={self.clicks}, CTR={ctr:.3f}")
    
    @property
    def ctr(self):
        return self.clicks / self.views if self.views > 0 else 0

class ABTestingServer:
    """
    Simplified Flask server logic
    """
    def __init__(self):
        self.bandit_A = ProductionBandit('Ad A')
        self.bandit_B = ProductionBandit('Ad B')
    
    def get_ad(self):
        """Endpoint: GET /get_ad"""
        # Thompson Sampling: ‡πÄ‡∏•‡∏∑‡∏≠‡∏Å arm ‡∏ó‡∏µ‡πà sample ‡πÑ‡∏î‡πâ‡∏™‡∏π‡∏á‡∏Å‡∏ß‡πà‡∏≤
        if self.bandit_A.sample() > self.bandit_B.sample():
            ad = 'A'
            self.bandit_A.add_view()
        else:
            ad = 'B'
            self.bandit_B.add_view()
        
        return ad
    
    def click_ad(self, ad_id):
        """Endpoint: POST /click_ad"""
        if ad_id == 'A':
            self.bandit_A.add_click()
        elif ad_id == 'B':
            self.bandit_B.add_click()
    
    def get_stats(self):
        """Get current statistics"""
        return {
            'A': {'clicks': self.bandit_A.clicks, 'views': self.bandit_A.views, 'ctr': self.bandit_A.ctr},
            'B': {'clicks': self.bandit_B.clicks, 'views': self.bandit_B.views, 'ctr': self.bandit_B.ctr}
        }

def simulate_client(server, true_ctr_A, true_ctr_B, n_users):
    """
    Simulate client.py behavior
    
    Args:
        server: ABTestingServer instance
        true_ctr_A: True click rate for Ad A (unknown to server)
        true_ctr_B: True click rate for Ad B (unknown to server)
        n_users: Number of users to simulate
    """
    print(f"\nüåê Simulating {n_users} users with production API...\n")
    print(f"True CTRs (unknown to server): A={true_ctr_A:.1%}, B={true_ctr_B:.1%}\n")
    
    selections = []
    
    for i in range(n_users):
        # 1. GET /get_ad
        ad = server.get_ad()
        selections.append(ad)
        
        # 2. Simulate user behavior (click or not?)
        true_ctr = true_ctr_A if ad == 'A' else true_ctr_B
        clicked = (np.random.random() < true_ctr)
        
        # 3. POST /click_ad (if clicked)
        if clicked:
            server.click_ad(ad)
    
    return selections

# Run simulation
print("\n" + "="*70)
print("üöÄ PRODUCTION A/B TESTING SIMULATION")
print("="*70)

# Initialize server
server = ABTestingServer()

# True CTRs (server ‡πÑ‡∏°‡πà‡∏£‡∏π‡πâ!)
TRUE_CTR_A = 0.08  # Ad A: 8% CTR
TRUE_CTR_B = 0.12  # Ad B: 12% CTR (better!)

# Simulate 500 users
selections = simulate_client(server, TRUE_CTR_A, TRUE_CTR_B, n_users=500)

# Final statistics
print("\n" + "="*70)
print("üìä FINAL RESULTS")
print("="*70 + "\n")

stats = server.get_stats()

for ad_id in ['A', 'B']:
    s = stats[ad_id]
    true_ctr = TRUE_CTR_A if ad_id == 'A' else TRUE_CTR_B
    print(f"Ad {ad_id}:")
    print(f"  Views: {s['views']}")
    print(f"  Clicks: {s['clicks']}")
    print(f"  Measured CTR: {s['ctr']:.3f} ({s['ctr']*100:.1f}%)")
    print(f"  True CTR: {true_ctr:.3f} ({true_ctr*100:.1f}%)")
    print(f"  Error: {abs(s['ctr'] - true_ctr)*100:.2f}%")
    print()

# Selection distribution
count_A = selections.count('A')
count_B = selections.count('B')

print(f"Selection Distribution:")
print(f"  Ad A: {count_A} times ({count_A/len(selections)*100:.1f}%)")
print(f"  Ad B: {count_B} times ({count_B/len(selections)*100:.1f}%)")

# Calculate regret
total_clicks = stats['A']['clicks'] + stats['B']['clicks']
optimal_clicks = len(selections) * TRUE_CTR_B  # If always chose B
regret_clicks = optimal_clicks - total_clicks

print(f"\nüí° Performance Analysis:")
print(f"  Total Clicks: {total_clicks}")
print(f"  Optimal (always B): {optimal_clicks:.1f}")
print(f"  Regret: {regret_clicks:.1f} clicks ({regret_clicks/optimal_clicks*100:.1f}% loss)")

print(f"\n‚úÖ Key Observations:")
print(f"  1. Algorithm learned that B is better!")
print(f"  2. Started equal ‚Üí Shifted to B over time")
print(f"  3. Still explores A occasionally (good!)")
print(f"  4. Regret is low (~{regret_clicks/optimal_clicks*100:.1f}% vs 50% for equal split)")

# Visualize selection over time
print(f"\nüìà Selection Evolution:")
window = 50
for i in range(0, len(selections), window):
    segment = selections[i:i+window]
    pct_B = segment.count('B') / len(segment) * 100
    bar = '‚ñà' * int(pct_B / 5)
    print(f"  Users {i+1:3d}-{min(i+window,len(selections)):3d}: B selected {pct_B:5.1f}% {bar}")
````
